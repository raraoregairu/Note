二次関数$y = -2 x^2 + 4 x + 1$の頂点を求め，グラフをかけ．

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

e = 1
h = -8
g = 11


# 頂点 (p, q) の座標を求める　
x, a, p, q = sympy.symbols('x a p q')

fx = e * x ** 2 + h * x + g

v = sympy.solve( a * (x - p) ** 2 + q - fx, [p, q, a])
p = v[p]
q = v[q]
print(f'頂点は ({p}, {q}) で、グラフは以下の通りである。')

# グラフを描く
xRange=(-, 2)                     # グラフ描画範囲
f = lambda x : e * x ** 2 + h * x + g # 関数

x = np.linspace(xRange[0], xRange[1], 100)
y = f(x)

plt.figure(dpi = 120)
plt.plot(x, y, color = '0.0')
plt.plot(p, q, marker = 'o')
plt.grid(color = '0.8')
plt.xlim(-60, 60)
plt.ylim(-60, 60)
plt.show()
```

2次関数$y = x ^ 2 + x - 6$について以下の問いに答えよ．
 1. $y = x ^ 2 + x - 6$のグラフをかけ．
 2. 二次不等式を$y = x ^ 2 + x - 6 < 0$を満たす$x$の範囲を求めよ．

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

# 頂点 (p,q) の座標を求める　
x, a, p, q = sympy.symbols('x a p q', real = True)

fx = x ** 2 + x - 6  # 関数

v = sympy.solve(a * (x - p) ** 2 + q - fx, [p, q, a])
p = v[p]
q = v[q]

print(f'(1) 頂点は ({p}, {q}) で、', end = '' )


# x軸との交点を求める
sol = sympy.solve( fx )
print("平方完成", f'{a}(x - {p}) ^ 2 + {q} ')
if len(sol) == 0 : # 実数解が存在しない
    print('x軸とは交わらない。', end = '' )
else :
    tx = ' と '.join(list(map(lambda x: '({0}, 0)'.format(x), sol)))
    print(f'x軸との交点は {tx} となる。', end = '' )

print('グラフは以下の通りである')

# グラフを描く
xRange=(-4 ,4)                # グラフ描画範囲
f = lambda x : x ** 2 + x - 6  # 関数

x = np.linspace(xRange[0], xRange[1], 100)
y = f(x)

plt.figure(dpi = 120)
plt.plot(x, y, color = '0.0')
plt.plot(p, q, marker = 'o')

if len(sol) != 0 :
    plt.hlines([0], xRange[0], xRange[1], linewidth = 0.75 )
    for s in sol :
        plt.plot(s, 0, marker = 'o', color = '0.0')    
    plt.fill_between(x, y, 0, where = y < 0, facecolor = 'b', alpha = 0.2)

plt.vlines([p], -10, 15)
plt.grid(color = '0.8')
plt.show()

print('(2) 不等式を満たす xの範囲は、',end = '')
if len(sol) == 2 :
    if sol[0] > sol[1] :
        print(f'{sol[1]} < x < {sol[0]} である。')
    else :
        print(f'{sol[0]} < x < {sol[1]} である。')
else:
    print('存在しない')
```

灰原哀

「ビルの間は５０メートル．飛び移るということになれば，６０メートルってところね．

隣のビルとの高低差は２０メートル．地球上の物体は，空間では水平方向に同じ速度で進むけど，

下向きには，重力により，決まった割合で速度を上げながら落下していくの」

「つまり，２０メートル落下するのに約２秒．２秒で６０メートル進まなくてはならない．

ということは，１秒間で３０メートル．時速に置き換えると１０８km/h」

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

# 頂点 (p, q) の座標を求める　
x, a, p, q = sympy.symbols('x a p q')

fx = -0.5 * 9.8 * x ** 2

v = sympy.solve( a * (x - p) ** 2 + q - fx, [p, q, a])
p = v[p]
q = v[q]
print(f'頂点は ({p}, {q}) で、グラフは以下の通りである。')

# グラフを描く
xRange=(0, 4)                     # グラフ描画範囲
f = lambda x : -0.5 * 9.8 * x ** 2  # 関数
fx = lambda x : 30 * x

x = np.linspace(xRange[0], xRange[1], 100)
x1 = fx(x)
y = f(x)

fig, ax1 = plt.subplots()
ax1.plot(x1, y, color = '0.0')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.grid(color = '0.8')

fig, ax2 = plt.subplots()
ax2.plot(x, y, color = '0.0')
ax2.set_xlabel("t")
ax2.set_ylabel("y")
ax2.grid(color = '0.8')

plt.show()
```
