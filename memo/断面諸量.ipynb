## 7. 断面諸量

### 7.1 断面０次モーメント

図形は様々な情報を持っている．

図形の断面積，図心位置，曲がりにくさが挙げられる．

これらの情報は**断面ｎ次モーメント**を使えばすべての値を求めることができる．

また，断面ｎ次モーメントは積分を使う必要があるが，規則性がある図形（三角形，長方形，円など）の場合は公式を使って簡単に計算できる．

まずは，下図に示すような直交座標系yz軸にある任意の図形を考える．

断面ｎ次モーメントは次式で表わされる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        J_{yn} = \int_{z1}^{z2} z^n b(z) dz = \int_A z^n dA
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        J_{zn} = \int_{y1}^{y2} y^n h(y) dy = \int_A y^n dA
    \end{align}
$$
    
</span>

このとき，$J_{yn}$はy軸周りの断面n次モーメント，$J_{zn}$はz軸周りの断面n次モーメント，nは次数．

n = 0を**断面０次モーメント**，n = 1を**断面１次モーメント**，n = 2を**断面２次モーメント**という．

断面０次モーメントは**断面積**を求めることができる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        J_{y0} = \int_A z^0 dA = \int_A dA = A
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        J_{z0} = \int_A y^0 dA = A
    \end{align}
$$
    
</span>

### 7.2 断面１次モーメント

**断面１次モーメント**は図形の**図心位置**を計算するときに用いられる物理量であり，断面積に距離をかけることで求められる．

単位は$[m^3]$となるが，これは体積ではなく，あくまで面積と距離の積を表わす．

定義式は次式によってあらわされる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_y = J_{y1} = \int_{z1}^{z2} z b(z) dz = \int_A z dA
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_z = J_{z1} = \int_{y1}^{y2} y h(y) dy = \int_A y dA 
    \end{align}
$$
    
</span>

このとき，$G_y$はy軸における断面１次モーメント$[m^3]$，$G_z$はz軸における断面１次モーメント$[m^3]$.

この断面１次モーメントの式を使って図心位置を求める．

図心位置の式を導出するときは座標軸と図心軸を用意し，図心軸での断面１次モーメントを計算する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_Y = \int_A (z - z_0) dA = \int_A zdA - \int_A z_0 dA = G_y - z_0 A
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_Z = \int_A (y - y_0) dA = \int_A y dA - \int_A y_0 dA = G_z - y_0 A
    \end{align}
$$
    
</span>

このとき，$G_Y$はY軸における断面１次モーメント$[m^3]$，$G_Z$はZ軸における断面１次モーメント$[m^3]$，$y_0$と$z_0$は図心位置$[m]$．

図心は一般的に回転が生じない点，図形を支えられる点などと言われているが，厳密にいえば断面１次モーメントが０になる点である．

断面１次モーメントが０となるためにアンバランスさによる回転が起きず，その１点で支えることができる．

図心軸の断面１次モーメント$G_y$と$G_z$に０を代入する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_y - z_0 A = 0 \qquad z_0 = \cfrac{G_y}{A}
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_z - y_0 A = 0 \qquad y_0 = \cfrac{G_z}{A}
    \end{align}
$$
    
</span>

図心位置は座標軸での断面１次モーメントを断面積で割れば求まる．

図心は重さを考慮していないのに対し，重心は重さを考慮している．

重さが均一の物体は図心と重心が等しくなる．

長方形の図心位置

断面１次モーメントと断面積は

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_y = \int_A z dA = \int_0^h zbdz = b \left[ \cfrac{z^2}{2} \right]_0^h = \cfrac{bh^2}{2}
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_z = \int_A y dA = \int_0^h ybdz = b \left[ \cfrac{y^2}{2} \right]_0^b = \cfrac{b^2h}{2}
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        A = bh
    \end{align}
$$
    
</span>

よって図心位置は

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        z_0 = \cfrac{G_y}{A} = \cfrac{bh^2 / 2}{bh} = \cfrac{h}{2}
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        y_0 = \cfrac{G_z}{A} = \cfrac{b^2h / 2}{bh} = \cfrac{b}{2}
    \end{align}
$$
    
</span>

長方形の図心位置は高さの半分，幅の半分である．

三角形のz軸方向の図心位置

相似比を使って$b'$を求め，そして微小面積を求める．

このとき，微小面積は長方形として扱う．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        b : b' = h : h - z
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        b' = \cfrac{b}{h}(h - z), \qquad dA = b'dz = \cfrac{b}{h}(h - z)dz
    \end{align}
$$
    
</span>

よって，断面１次モーメント，断面積，図心位置は

</span>

<span style="font-size:120%;">
$$
    \begin{align}
        G_y = \int_A z dA = \int_0^h \cfrac{b}{h}(hz - z^2)dz = \cfrac{b}{h} \left[ \cfrac{h z^2}{2} - \cfrac{z^3}{3} \right]_0^h = \cfrac{bh^2}{6}
    \end{align}
$$
    
</span>

</span>

<span style="font-size:120%;">
$$
    \begin{align}
        A = \cfrac{1}{2} bh z_0 = \cfrac{G_y}{A} = \cfrac{b h^2 / 6}{1 / 2 bh} = \cfrac{h}{3}
    \end{align}
$$
    
</span>

三角形のz軸方向の図心位置は高さの1 / 3倍にあることがわかる．

半円形のz軸方向の図心位置

微小面積は

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        b = \cfrac{d}{2} cos \theta = d cos \theta
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        z = \cfrac{d}{2} sin \theta
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        dz = \cfrac{d}{2} cos \theta d \theta
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        dA = bdz =d cos \theta dz = \cfrac{d^2}{2} cos^2 \theta d \theta
    \end{align}
$$
    
</span>

断面１次モーメントは

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_y &= \int_A  z dA \\
            &= \int_0^{d / 2} z dA \\
            &= \cfrac{d^3}{4} \int_0^{\pi / 2} cos^2 \theta sin\theta d \theta
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_y &= \cfrac{d^3}{4} \int_0^{\pi / 2} (1 - sin^2 \theta) sin\theta d \theta \\
            &= \cfrac{d^3}{4} \int_0^{\pi / 2} sin\theta - sin^3 \theta d \theta \\
            &= \cfrac{d^3}{4} \int_0^{\pi / 2} sin\theta d \theta - \cfrac{d^3}{4} \int_0^{\pi / 2} (1 - cos^2 \theta)sin\theta d \theta
    \end{align}
$$
    
</span>

$(sin \theta)^3$を積分するには，三倍角の公式もしくは置換積分を使う必要がある．

置換積分では

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        t &= cos \theta \\
        dt &= -sin \theta d \theta
    \end{align}
$$
    
</span>

上式をy軸の断面１次モーメントの式に代入し，積分を続ける．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        G_y &= \cfrac{d^3}{4} \int_0^{\pi / 2} sin\theta d \theta - \cfrac{d^3}{4} \int_0^{\pi / 2} (1 - cos^2 \theta)sin\theta d \theta \\
            &= \cfrac{d^3}{4} \int_0^{\pi / 2} sin\theta d \theta - \cfrac{d^3}{4} \int_0^{\pi / 2} - (1 - t ^ 2) dt \\ 
            &= \cfrac{d^3}{4} \left[ - cos \theta \right]_0^{\pi / 2} - \cfrac{d^3}{4} \left[t - \cfrac{t^3}{3} \right]_0^1 \\
            &= \cfrac{d^3}{4} (0 + 1) - \cfrac{d^3}{4} (1 - \cfrac{1}{3} - 0) \\
            &= \cfrac{d^3}{12} 
        A   &= \left( \cfrac{d}{2} \right)^2 \cdot \cfrac{1}{2} \cdot \pi = \cfrac{\pi d^2}{8}
    \end{align} 
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        A  = \left( \cfrac{d}{2} \right)^2 \cdot \cfrac{1}{2} \cdot \pi = \cfrac{\pi d^2}{8}            
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        z_0 = \cfrac{d^3 / 12}{\pi d^2 / 8} = \cfrac{2d}{3 \pi}          
    \end{align}
$$
    
</span>

z軸方向の図心位置は$2d / 3 \pi $であり，ちなみにy軸方向の図心位置は対称性より$d / 2$の位置にある．

### 7.3 断面２次モーメント

一般的に断面積が大きいと部材は曲がりにくくなる．

しかし，断面積が同じであっても断面形状によっては曲がりやすい場合もある．

これら部材の曲がりにくさを数値化したものを**断面２次モーメント**という．

断面２次モーメントは$面積 \times 長さ^2$で計算することができ，単位は$[m^4]$となる．

この値を求めることで断面形状による変化のしにくさを求めることができる．

定義式は次式によってあらわされる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_y = J_{y2} = \int_{z1}^{z2} z^2 b(z) dz = \int_A z^2 dA
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_z = J_{z2} = \int_{y1}^{y2} y^2 h(y) dy = \int_A y^2 dA
    \end{align}
$$
    
</span>

このとき，$I_y$はy軸における断面２次モーメント$[m^4]$，$I_z$はz軸における断面弐次モーメント$[m^4]$.

定義式から断面２次モーメントはどのような場合でも**正値**を示すことがわかる．

断面諸量の解き方は以下のとおりである

 1. 座標軸をy軸・z軸とし，断面１次モーメントから図心位置を求める

 2. 図心軸をy軸・z軸とし，図心位置での断面２次モーメント$I_y$，$I_z$を求める．

 3. 図心位置以外の断面２次モーメントを次式から求める．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_Y &= \int_A (z - z_1)^2 dA \\
            &= \int_A z^2 dA - 2z_1 \int_A z dA + z_1^2 \int_A dA \\
            &= I_y - 2 z_1 G_y + z_1^2 A
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_Z &= \int_A (y - y_1)^2 dA \\
            &= \int_A y^2 dA - 2y_1 \int_A y dA + y_1^2 \int_A dA \\
            &= I_z - 2 y_1 G_z + y_1^2 A
    \end{align}
$$
    
</span>

このとき，図心位置での断面１次モーメント$G_y$，$G_z$は0になるので$I_Y$，$I_Z$は次式で表わすことができる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_Y &= I_y + z_1^2 A \\
        I_Z &= I_z + y_1^2 A
    \end{align}
$$
    
</span>

断面２次モーメントは**図心軸を中心に回転させたときの抵抗係数**を表している．

そのため，図心から離れるほど断面２次モーメントの値は大きくなる．

この値が間接的に曲げモーメントの抵抗係数（変形しにくさ）を表している．

長方形の図心軸（y軸，z軸）における断面２次モーメント，Y軸，Z軸の断面２次モーメント．

図心軸であるy軸，z軸の断面２次モーメント

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_y &= \int_A z^2 dA = \int_{- h / 2}^{h / 2} z^2 b dz = \left[ \cfrac{b z^3}{3}\right]_{- h / 2}^{h / 2} \\
            &= \cfrac{b}{3} \left( \cfrac{h}{2} \right)^3 - \cfrac{b}{3} \left( - \cfrac{h}{2} \right)^3 = \cfrac{b h^3}{12}
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_z &= \int_A y^2 dA = \int_{- b / 2}^{b / 2} y^2 h dy = \left[ \cfrac{h y^3}{3}\right]_{- b / 2}^{b / 2} \\
            &= \cfrac{h}{3} \left( \cfrac{b}{2} \right)^3 - \cfrac{h}{3} \left( - \cfrac{b}{2} \right)^3 = \cfrac{b^3 h}{12}
    \end{align}
$$
    
</span>

任意軸であるY軸，Z軸の断面２次モーメントは

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_Y &= I_y + z_1^2 A = \cfrac{b h^3}{12} + \left( \cfrac{h}{2} \right)^2 bh \\
            &= \cfrac{b h^3}{12} + \cfrac{b h^3}{4} = \cfrac{b h^3}{3}
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_Z &= I_z + y_1^2 A = \cfrac{b^3 h}{12} + \left( \cfrac{b}{2} \right)^2 bh \\
            &= \cfrac{b^3 h}{12} + \cfrac{b^3 h}{4} = \cfrac{b^3 h}{3}
    \end{align}
$$
    
</span>

二等辺三角形の図心軸における断面２次モーメント

比を使って微小面積を求め，y軸の断面２次モーメントを計算する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &b : b' = h : \cfrac{2}{3} h - z \\
        &b' = \cfrac{b}{h} \left( \cfrac{2}{3} h - z \right) \\
        &dA = b'dz
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_y &= \int_A z^2 dA = \int_{- h / 3}^{2h / 3} z^2 \cfrac{b}{h} \left( \cfrac{2}{3} h - z \right) dz \\
            &= \cfrac{b}{h} \left[ \cfrac{2}{9} h z^3 - \cfrac{z^4}{4} \right]_{-h / 3}^{2h / 3} \\
            &= \cfrac{16 b h^3}{243} - \cfrac{16 b h^3}{324} - \left( - \cfrac{2 b h^3}{243} - \cfrac{b h^3}{324} \right) \\
            &= \cfrac{b h^3}{36}
    \end{align}
$$
    
</span>

z軸の断面２次モーメントは

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &h : h' = \cfrac{b}{2} : \cfrac{b}{2} - y \\
        &h' = \cfrac{2h}{b} \left( \cfrac{b}{2} - y \right) \\
        &dA = h'dy
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_z &= \int_A y^2 dA = \int_{- b / 2}^{b / 2} y^2 \cfrac{2h}{b} \left( \cfrac{b}{2} - y \right) dy \\
            &= \cfrac{4h}{b} \left[ \cfrac{b y^3}{6} - \cfrac{y^4}{4} \right]_{0}^{b / 2} \\
            &= \cfrac{b^3 h}{12} - \cfrac{b^3 h}{16} \\
            &= \cfrac{b^3 h}{48}
    \end{align}
$$
    
</span>

円形のy軸の断面２次モーメント

三角関数を使って微小面積を求める．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        b &= \cfrac{d}{2} cos \theta \times 2 = d cos \theta \\
        z &= \cfrac{d}{2} sin \theta , dz = \cfrac{d}{2} cos \theta d \theta \\
        dA &= bdz = d cos \theta dz = \cfrac{d^2}{2} cos^2 \theta d \theta
    \end{align}
$$
    
</span>

y軸の断面２次モーメントを求める．このとき，三角関数の倍角の公式を使用する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        I_y &= \int_A z^2 dA = \int_{- d / 2}^{d / 2} z^2 dA \\
            &= 2 \int_{0}^{d / 2} z^2 dA = \cfrac{d^4}{4} \int_0^{\pi / 2} cos^2 \theta sin^2 \theta d \theta \\
            &= \cfrac{d^4}{16} \int_0^{\pi / 2} sin^2 2 \theta d \theta \\
            &= \cfrac{d^4}{16} \int_0^{\pi / 2} \cfrac{1 - cos 4 \theta}{2} d \theta \\
            &= \cfrac{d^4}{16} \left[\cfrac{1}{2} \theta - \cfrac{sin 4 \theta}{8} \right]_0^{\pi / 2} = \cfrac{\pi d^4}{64}
    \end{align}
$$
    
</span>
