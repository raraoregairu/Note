## Markdownとは
  
---

* アメリカ合衆国，フィラデルフィア出身のライター・コラムニストであるジョン・グルーバー(John Gruber)によって  
  2004年に開発された「フォーマット」である．
<br>

* デジタル文章を作るときに使うHTML(HyperText Markup Language)などの<u>マークアップ言語</u>を簡略化  
  できるようにプログラムされた．
  
---  
  
<details><summary><span style="font-size: 80%;">▶マークアップ言語とは</span></summary>

Webサイトの文章・文字のサイズ・画像などにタグ付けをし，コンピュータに伝達するための言語
    
「タグ」という特定の文字列で囲むことで，元の文に構造・装飾などを反映させる言語

https://bolt-dev.net/posts/440
    
</details>

---

### Markdownの特徴：

---

    1. 手軽に文章構造を明示できる
    2. 簡単で，覚えやすい
    3. 読み書きに特別なアプリ
    4. それでいて，対応アプリを使えば快適に読み書きできる

[参照](https://backlog.com/ja/blog/how-to-write-markdown/) 

## Jupyter_NotebookにおけるMarkdown

---
Markdown（HTMLを含む）についてまとめていく．

### 改行

---

#### ◎Markdown :

    行末で半角スペースを二個連続入力して改行

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
テキスト  /n
テキスト２
```

▶実行
    
テキスト  
テキスト２

</details>

---

#### ◎ HTML :

    brタグをかく

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
テキスト<br>テキスト２
```

▶実行

テキスト<br>テキスト２

</details>

---

### 見出し

---

#### ◎Markdown :

    #（シャープ）＋ 半角スペース ＋ テキスト，# の個数で見出しの大きさが変わる

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
# テキスト
## テキスト
```

▶実行

# テキスト
## テキスト

</details>

---

#### ◎HTML :

    hタグで挟む hタグにはh1からh6タグまであり，数値で見出しに大きさが変わる．

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```       
<h1>テキスト</h1>
<h2>テキスト</h2>
```

▶実行

<h1>テキスト</h1>
<h2>テキスト</h2>

</details>

---

### 箇条書き

---

#### ◎Markdown : 

    * (アスタリスク) or + (プラス) or - (ハイフン)の後に半角スペース

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
* テキスト1
+ テキスト2
- テキスト3
```

▶実行

* テキスト1
+ テキスト2
- テキスト3

</details>

---

#### ◎HTML :

    テキストをliタグで挟み，liタグをulタグで挟む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<ul>
  <li>テキスト1</li>
  <li>テキスト2</li>
</ul>
```

▶実行

<ul>
  <li>テキスト1</li>
  <li>テキスト2</li>
</ul>

</details>

---

### 段落番号

---

#### ◎Markdown : 

    1. テキスト　数値＋半角ドット＋半角スペース

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
1. テキスト1
2. テキスト2
```

▶実行

1. テキスト1
2. テキスト2

</details>

---

#### ◎HTML :

    テキストをliタグで挟み，liタグをolタグで挟む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<ol>
  <li>テキスト1</li>
  <li>テキスト2</li>
</ol>
```

▶実行

<ol>
  <li>テキスト1</li>
  <li>テキスト2</li>
</ol>

</details>

---

#### チェックボックス

---

#### ◎Markdown : 

    - （ハイフン）＋[]（角括弧）＋ 半角スペース　角括弧内のxによりチェックの有無

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
- [ ] テキスト1
- [x] テキスト2
```

▶実行

- [ ] テキスト1
- [x] テキスト2

</details>
---

#### ◎HTML :

    inputタグの中のtype属性にcheckboxを設定する．タグ内のcheckedによりチェックの有無

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<input type="checkbox"> テキスト1

<input type="checkbox" checked> テキスト2
```

▶実行

<input type="checkbox"> テキスト1  
    
<input type="checkbox" checked> テキスト2

</details>

---

### 太字

---

#### ◎Markdown : 

    *（アスタリスク）２つもしくは_（アンダースコア）２つでテキストを囲む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
**テキスト**  
__テキスト__
```

▶実行

**テキスト**  
__テキスト__

</details>

---

#### ◎HTML :

    strongタグでテキストを挟む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<strong>テキスト</strong>
```

▶実行

<strong>テキスト</strong>

</details>

### 斜体

---

#### ◎Markdown : 

    *（アスタリスク）１つもしくは_（アンダースコア）１つでテキストを囲む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
*text1*  
_text2_
```

▶実行

*text1*  
_text2_

</details>

#### ◎HTML :

    iタグでテキストを囲む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<i>text</i>
```

▶実行

<i>text</i>

</details>

### 下線

---

#### ◎HTML :

    uタグでテキストを囲む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<u>テキスト</u>
```

▶実行

<u>テキスト</u>

</details>

---

### 打消し線

---

#### ◎Markdown : 

    ~ （チルダ）２つでテキストを囲む

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
~~テキスト~~
```

▶実行

~~テキスト~~

</details>

---   

#### ◎HTML :

    sタグをかく

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<s>テキスト</s>
```

▶実行

<s>テキスト</s>

</details>

### 水平線

---

#### ◎Markdown : 

    _（アンダースコア），*（アスタリスク），-（ハイフン）を３つ以上をつける

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
***
---
+++
```

▶実行

***
---
+++

</details>    

#### ◎HTML :

    hrタグをかく

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
<hr>
```

▶実行

<hr>

</details>

### ページ内リンク


[見出し1](#見出し1)
.,が入っている場合()内では無視して記述
スペースの場合ハイフンに置き換えて記述
始まりが大文字の場合小文字に置き換えて記述

定義参照リンク
[こっちからgoogle][google]
その他の文章
[こっちからもgoogle][google]

[google]: https://www.google.co.jp/

ハイパーリンク
[リンクテキスト](URL)

画像
![代替テキスト](画像URL)

引用
> テキスト

pre記法
半角スペース4もしくはタブ

code記法
`バッククォートで文字列を囲む（インラインコード）`

インラインコード
プログラムの中でほかのコードと同じ行に係れること

```バッククォート三つ（コードブロック）
```

テーブル
「|」は列の区切りに使われる

空白を入れずに「|」入れることでその左のセルを結合することができる

（横にセルを連結することはできるが、縦に連結することはできない。）

複数の「-」はヘッダー部とボディ部の区切りに使われる。また、その両端を「:」で囲むことでその列を中央寄せに、右端に「:」を付けることで右寄せになる。

最後に「[]」で囲んだ文字列はテーブルのタイトルになる。

|                 |             結合               ||
|ヘッダー１| ヘッダー２|ヘッダー３|
| :------------- | :------------: | -------------: |
|セル1        |      長いセル               ||
|セル2        |  中央寄せ  |  右寄せ     |
|セル3        |  **太字**  |  *syatai*     |
|セル4        |          さらに               ||
