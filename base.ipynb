# 1. 水力学

## 1.1 水の性質

### 1.1.1. 水の物理量

全ての物体には，質量がある．水にも質量が存在しており，1[L]で1[kg]となる．

また，質量に重力加速度をかけたものを重力または**重量**とよぶ．

<br>
<span style="font-size:150%;">
$$
    \begin{align}
        W = mg
    \end{align}
$$
    
</span>
<br>

このとき，Wは重量[N], m は質量[kg], gは重力加速度[$m / s ^ 2$]

物質の重さを比較する方法として**密度**がある．水の密度は1000[$kg/m^3$]であり，1[$m^3$]の容器に水を入れると1000[kg]である．

実際は，温度や気圧により体積が変化するため，水力学ではほとんどの場合無視する．

<br>
<span style="font-size:150%;">
$$
    \begin{align}
        \rho = m / V
    \end{align}
$$
    
</span>
<br>

このとき，$\rho$は密度[$kg / m^3$], Vは体積[$m^3$]．

ちなみに，海水の密度は1030[$kg / m^3$], 油は720[$kg / m^3$]，エタノールは790[$kg / m^3$]，水銀は1400[$kg / m^3$]くらい．海水が水より重いのは酸素や水素のほかにナトリウムやマグネシウムなどの様々なものが混ざっているから．また，油が水に浮くのは，油の密度が水より小さいから．比較することで物体の特性がより見える．

密度以外の比較方法として**単位体積重量**がある．土質力学でよく使う．水の単位体積重量は9800[$N / m^3$]である．

<br>
<span style="font-size:150%;">
$$
    \begin{align}
        \gamma = W / V = mg /V = \rho g
    \end{align}
$$
    
</span>
<br>

このとき，$\gamma$は単位体積重量[$N / m^3$]

ほかの比較方法としては**比重s**が挙げられる．比重は物体の密度を水の密度で割った値であり，単位は存在しない．水の比重は1.00

<br>
<span style="font-size:150%;">
$$
    \begin{align}
        s = \rho / 1000
    \end{align}
$$
    
</span>
<br>

### 1.1.2 水の表面張力

ざらざらの机の上に水を一滴落とすと水は広がらずに球形を保つ．

水分子はバラバラにならないようにお互いに強い力に引っ張り合う．逆に，机や空気と触れている水分子は引っ張り合う相手がいないため，少し不安定な状態になる．

不安定な状態の分子は安定状態になるために表面積を小さくしようとする．

その表面積を小さくしようとする力を**界面張力**という．液相/気相間での場合は特に**表面張力**．

表面張力があるために水は机上で広がらずに球体を保ち続ける．


<br>
<span style="font-size:150%;">
$$
    \begin{align}
        \gamma_s = \gamma_l cos \theta + \gamma_{sl}
    \end{align}
$$
    
</span>
<br>

このとき，$\gamma_s$は固体の表面張力，$\gamma_l$は液体の表面張力，$\gamma_{sl}$は固体と液体の界面張力．

上式をヤングの式といい，右辺が大きいほど液体は球体に近づく．また，角度$\theta$が90°より小さいときを疎水性，90°より大きいときを親水性という．

撥水加工の製品（服や傘，水着など）は角度$\theta$を小さくしたり，界面張力を大きくするような加工がある．

一方で，表面張力を弱める方法がある．それは**乳化**という．

水と油は放っておいても混ざることはない．両方とも表面張力が強くそれぞれの分子同士で結合しているから．混ぜるためにはそれぞれの結合を弱める必要がある．

水と油は激しい振動を与えれば一時的に結合が弱くなる．（ドレッシングをかける前に振る）

乳化は時間が経つと水と油に分離する．時間経過によって分離させないものとして**界面活性剤**がある．

界面活性剤は親水基（水と結合する部分）と疎水基（油と結合する部分）の両方を持っており，水と油を混ぜ合わせた状態で保つ．

また，水に界面活性剤を入れて机上に一滴落とすと，球体を作らずに広がる．界面活性剤によって分子間の結合力が弱まるから．

界面活性剤は，身の回りの製品にたくさん使われる．

たとえば，石鹸．人間の体から出る汚れの大部分が油性．体に石鹼をつけて水に流すと，汚れと石鹸が結合し，体から流れ落ちる．

一方，泥汚れは石鹸で落とそうとしても効果がない．界面活性剤は化粧品やマヨネーズなど使われている．

### 1.1.3 水の毛細管現象

**毛細管現象**とは，液体中に細い管（毛細管）を入れると，管内の水面が管外の水面より高くなったり低くなったりする現象．

水の中にガラス管を入れると，管の円周付近の水面がほんの少し上昇する．

このガラス管の円周に発生する力は表面張力．また，鉛直方向に作用する表面張力の合力を**付着力**という．

毛細管現象は液体とガラス管の関係が重要．水とガラス管は親水性で濡れようとして水面が上昇する．

一方で，水銀とガラス管が疎水性なので水を弾こうとして水面は下降する．

水面上昇によって生じた曲面を*メニスカス*と呼ぶ．水面が上昇したとき凹型メニスカス，水面が下降したとき凸型メニスカスを形成する．

一般的に，凹型メニスカスは下面の目盛りを読み，凸型メニスカスは上面の目盛りを読む．

表面張力による水の上昇量を力のつり合いより求める．

水面が最高点まで上昇したとする．鉛直方向に働いている力は付着力と水の重量．

実際には，管内の水面に大気圧が作用しているのだが，**パスカルの原理**により同じ大きさの圧力が管内の下面にも作用する．

パスカルの原理とは「密閉された容器に圧力をかけるとその容器の形に関係なく他のすべての部分に伝える」という基本原理．

そのため，大気圧は無視することができ，重量と付着力のつり合い式から表面張力による水の上昇量が求まる．

ヤングーラプラスの式

水が高さ$h$変動した時，密度$\rho$の水の重量は$\rho \cdot g \cdot (h \times \pi (D / 2)^2)$

吸着力は水と接するガラス管内面の長さと表面張力の積より$\pi D \times T cos \theta$

<span style="font-size:150%;">
$$
    \begin{align}
        \rho g \cfrac{\pi D^2}{4} h = \pi DT cos \theta
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        h = \cfrac{4Tcos \theta}{\rho g D}
    \end{align}
$$
    
</span>

このとき，$D$は管の直径[m]，$h$は水の高さ[m]，$T$は表面張力[N/m]．

液体中に細い管を入れると管内の液体表面が上下に変動する．毛細管現象は水が持つエネルギーを測定するのに役立つ．

### 1.1.4 水の圧縮性

まずは，空気の圧縮性について

人間は5[m]ほど海に潜ると耳に激痛が走り始める．これは，耳の中の空気が水圧によって圧縮するため．

空気は圧力がかかると体積を変化させる．そのため，海に深く潜るときは入水前に必ず耳の中の空気を抜かないといけない

気体の体積と圧力の関係は**気体の状態方程式**によってあらわされる．

<span style="font-size:150%;">
$$
    \begin{align}
        pV = nRT
    \end{align}
$$
    
</span>

このとき$n$はモル数[mol]，$R$は気体の状態定数8.314[J/(K・mol)]，$T$は温度[K].

圧力と体積は反比例の関係にあり，圧力が大きくなれば空気は圧縮する．

次に水の圧縮性についてだが，水は圧縮しない．この圧縮しない性質を**非圧縮性**とよぶ．

なぜ圧縮しないのかはまずマッハを知る必要がある．

マッハとは物体の速さを音が伝わる速さで割った値であり，式で表わすと次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        M = U / c
    \end{align}
$$
    
</span>

このとき，$M$はマッハ[単位なし]，$U$は物体の速さ[m/s]，$c$は音の伝わる速さ[m/s].

空気中での音が伝わる速さは340[m/s]なのでマッハ2では680[m/s]，マッハ3なら1020[m/s]である．

水での音が伝わる速さは1500[m/s]程度であり，水中をマッハ１の速度を出すには秒速1500[m/s]進まないといけない．

水中を最速で動けるのは原子力潜水艦だが，速度は20[m/s]である．

このマッハが圧縮性の指標となっている．ある流体中をマッハ1/3以上で動ける場合は圧縮性，1/3以上で動けない場合は非圧縮性と定義されている．

液体中の音速は**体積弾性率**を$K$として

<span style="font-size:150%;">
$$
    \begin{align}
        c = \sqrt{\cfrac{K}{\rho}}
    \end{align}
$$
    
</span>

￥体積弾性率$K[Pa]$は圧縮率の逆数である．

<span style="font-size:150%;">
$$
    \begin{align}
        K = \cfrac{1}{\beta}
    \end{align}
$$
    
</span>

圧縮率$\beta[m^2/N]$は圧力が上がったときに体積がどれだけ表わす係数である．

圧力変化$dP$が微小であれば，圧力変化$dP$と体積変化$dV$との間には以下の関係が成立する．

<span style="font-size:150%;">
$$
    \begin{align}
        dP = - K \cfrac{dV}{V}
    \end{align}
$$
    
</span>

ここで，$dV/V$は体積変化の割合を示す体積ひずみである．よって，

<span style="font-size:150%;">
$$
    \begin{align}
        \beta = - \cfrac{1}{V} \cfrac{dV}{dP} = \cfrac{1}{\rho} \cfrac{d \rho}{dP}
    \end{align}
$$
    
</span>

水の温度が20℃のとき，体積弾性係数は$K = 2.20 \times 10^9[Pa]$,密度は998.20$[kg/m^3]$となるため，音の伝わる速さは1485[m/s]となる．


### 1.1.5 水の粘性

**粘性**とは，**流体（気体と液体の総称）**が持つ性質であり，物体が速く動こうとするときに流体が抵抗する性質．

気体の粘性は，気体中に存在する速度の遅い分子と速い分子が衝突し，速度が平均化するのが原因である．

一方で，液体の粘性は，液体が形を変えようとするときに，隣の分子と離れないよう発生する分子間力が原因である．

粘性によって生じた応力を摩擦応力または粘性応力という．

応力とは単位面積あたりに作用する力のことであり，単位は圧力と同じ$[N/m^2]$または$[Pa]$を使う．

しかし，圧力とは考え方や力の作用位置が異なる．

圧力はある面に対して垂直に作用するが，摩擦応力は面に対して平行に作用する．

圧力との違いはもう一つある．それは，圧力は水が静止した状態でも作用するのに対し，摩擦応力は水が流れていないと作用しない．

流れ場（水が流れている空間）において速度差があれば摩擦応力が発生する．

壁は静止しているので，水が流れているとき壁面では必ず摩擦応力が作用する．

速度差がある水同士の距離が近いほど摩擦応力は強くなる．

この法則を**ニュートンの粘性法則**という．

ニュートンの粘性法則にしたがう流体をニュートン流体とよぶ．水はニュートン流体である．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{F}{A} = \tau = \mu \cfrac{du}{dz}
    \end{align}
$$
    
</span>

このとき，$F$は力[N]，$A$は面積[A]，$\tau$は摩擦応力[Pa]，$\mu$は粘性係数or粘度$[Pa \cdot s]$，$u$は速度[m/s]，$z$は距離[m]である．また，$du/dz$を速度勾配という．

**粘性係数**は流体の粘りの度合いを表しているが，$[Pa \cdot s]$という非常に複雑な単位をもつ．

そこで，この粘性係数を密度で割った物理量を**動粘性係数**または動粘度という．

<span style="font-size:150%;">
$$
    \begin{align}
        \nu= \cfrac{\mu}{\rho}
    \end{align}
$$
    
</span>

このとき，$\nu$は動粘性係数or動粘度$[m^2/s]$．

動粘性係数は流体の流れの乱れやすさの指標となる．

これは水分子が温度上昇により振動をし始め，その結果，水分子同士の結合力が弱くなり，流動性が高まるためである．

一方，空気の粘性係数（動粘性係数）は温度が上がると上昇する．

これは，温度が上昇したことにより分子同士の衝突回数が多くなり，速度の平均化が起きるためである．

また，空気の動粘性係数は，水の動粘性係数より10倍ほどおおきいことがわかる．

そのため，空気のほうが流れは乱れやすいといえる．

</span>

$$
    
</span>

</span>

    \end{align}
$$
    
</span>

このとき，$\tau_0$は降伏強度$[N/m^2]$，$\eta$は非ニュートン流体の粘性係数$[Pa \cdot s]$，$n$は次数[単位なし]である．

ビンガム流体は$\tau_0 > 0$で$n = 1$，ダイラタント流体は$\tau_0 = 0$で$n > 1$，擬塑性流体は$\tau_0 = 0$で$n < 1$．


#### ビンガム流体

ビンガム流体の例はとしては，歯磨き粉がある．

歯磨き粉の入ったチューブは逆さにしても出ない．

しかし，一度チューブに力を加えると止まることなく歯磨き粉が出てくる．

このように，力を加えると流動する流体をビンガム流体という．

粘性力と速度勾配は比例（一次関数）の関係にある．

しかし，切片が原点になく上のほうに存在する．

この切片を降伏強度といい，降伏強度以下だとビンガム流体は流動しない．

歯磨き粉以外だと，ケチャップや塗料がある．

#### ダイラタント流体

ダイラタント流体は生クリームやミルクチョコレート，波打ち際の砂などがあてはまる．

かき混ぜる前の生クリームは泡だて器ですくってもするすると落ちていく．

しかし，混ぜ続けると徐々に粘性が大きくなっていく．

このように，速度勾配が大きくなれば二次関数的に粘性力が大きくなる流体はダイラタント流体という．

#### 擬塑性流体

擬塑性流体は速度勾配が大きければ，粘性力は小さくなる．

例として，ポールペンインクがある．

ボールペンの芯を出した状態で逆さにしてもインクが漏れることはない．

動いていないときはインクの粘性力が大きいからである．

しかし，紙に文字を書き始めるとインクが簡単に出る.

これは速度勾配が増加し，粘性力が小さくなったためである．

他にマヨネーズや濃縮ジュースがある．

## 1.2 静水の力学

### 1.2.1 静水圧

静水とは動いていない水のことであり，ダムや貯水池の水が挙げられる．

また，海や湖も波を考えなければ静水となる．

静水における水圧を**静水圧**という．

静水圧を表現する方法として**ゲージ圧**と**絶対圧**が挙げられる．

ゲージ圧は大気圧を基準にしており，<u>地表面の圧力</u>を０気圧として扱う．

大気圧より高い圧力を「**正圧**」，低い圧力を「**負圧**」という．

そのため，ゲージ圧には負の圧力が存在する．

一方，絶対圧は真空状態を基準としており，負の圧力が存在しない．

水力学ではゲージ圧を使用するのがほとんどである．

また，面積全体にかかる静水圧を合計した力を**全水圧**という．

<span style="font-size:150%;">
$$
    \begin{align}
        P = \rho g h A
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p = \cfrac{P}{A} = \rho g h 
    \end{align}
$$
    
</span>

このとき，$P$は全水圧$[N]$，$A$は水圧が作用する面積$[m^2]$，$p$は静水圧$[N/m^2]$，$h$は水深$[m]$．

また，静水圧の式を変形すると**水深**（水の深さ）がもめられる．

密度と重力加速度を定数とすると，水深と水圧は比例の関係にあることがわかる．

また，圧力を単位体積重量で割ったものを**水頭**とよぶ．

<span style="font-size:150%;">
$$
    \begin{align}
        h = \cfrac{p}{\rho g} 
    \end{align}
$$
    
</span>

水深も水頭も同じ$[m]$の単位で扱われる．

しかし，水深は単に水面からの距離を表しているのに対し，水頭は水の持つエネルギーの大きさを表している．

通常，エネルギーの量は[J]が使われるが，水力学ではエネルギー量を水頭で表現し，図示することにより簡単にエネルギー量を認識できるようにしている．



大気圧が水を押し上げることのできる限界の高さは

<span style="font-size:150%;">
$$
    \begin{align}
        h = \cfrac{p}{\rho g}  = \cfrac{101.3 \times 1000}{1000 \times 9.8} = 10.33 m
    \end{align}
$$
    
</span>

海水による水圧がちょうど一気圧となる水深は

<span style="font-size:150%;">
$$
    \begin{align}
        h = \cfrac{p}{\rho g}  = \cfrac{101.3 \times 1000}{1030 \times 9.8} = 10.02 m
    \end{align}
$$
    
</span>

水圧計とは，測定したい部分に小さい穴（小孔）を開け，透明な管（細管）を通すことで水深を測定し，水圧を求める計測器である．

水圧計には，マノメーター，傾斜マノメーター，水銀マノメーターの３種類がある．

傾斜マノメーターは，管を傾斜させるために水が上昇するときの距離が長くなり，目盛りを細かく設定できる．

そのため，傾斜マノメーターは圧力が小さい時によく利用される．

逆に，水銀マノメーターは圧力が高いときに使われる．

それは，水銀の密度が水の約13.5倍大きいために，水の上昇量を抑えることができるからである．

傾斜マノメーターにおける圧力は次式で求められる．

<span style="font-size:150%;">
$$
    \begin{align}
        p = \rho g l sin \theta
    \end{align}
$$
    
</span>

また，水銀マノメーターの圧力は次式で表わされる．

<span style="font-size:150%;">
$$
    \begin{align}
        p + \rho g h_1 = \rho g h_2
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p = \rho g h_2 - \rho g h_1
    \end{align}
$$
    
</span>

マノメーターの原理を用いて２点間の圧力差を求めることができる．

圧力差を求めるときに使用する器具を**差圧計**と呼び，圧力差の式は次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        p_A + \rho g h_1 = p_B + \rho g h_2 + \rho' g h_3
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p_A - p_B = \rho g (h_2 - h_1) + \rho' g h_3
    \end{align}
$$
    
</span>

n-n断面より上の圧力は，液体と高さが同じであるため圧力も等しくなる．

n-n断面より上の圧力を$p_n$とすると次の式が成り立つ．$\rho'= 0.9$

<span style="font-size:150%;">
$$
    \begin{align}
        p_A = p_n + 0.9 \times 9.8 \times 1.2 + 1.0 \times 9.8 \times 0.6 = p_n + 16.46[kN / m^2]
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p_B = p_n + 1.0 \times 9.8 \times 0.8 = p_n + 7.84[kN / m^2]
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p_A - p_B = 16.84 - 7.84 = 8.62[kN/m^2]
    \end{align}
$$
    
</span>

静水圧の原理の一つに**パスカルの原理**がある．

パスカルの原理とは密閉された流体に圧力を加えると，流体の各部分に同じだけの圧力が伝えられるという原理である．

例：水枕の一部を強く押すと，表面のどの部分もビーンと張って同じような弾力を感ずる．

圧力はどの面に対しても垂直に作用している．

パスカルの原理を用いた道具として，水圧機や油圧ブレーキ（車のブレーキ）が挙げられる．

水圧機を簡単な図として表すと上記のようになる．

水圧機はパスカルの原理を使って小さい力で大きなものを動かす．

マノメーターを計算したときと同様に，n-n断面での圧力のつり合いから以下のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{P_1}{A_1} + \rho g h = \cfrac{P_2}{A_2}
    \end{align}
$$
    
</span>

このとき，面積にかかる力がものすごく大きいとすると，水圧$\rho g h$が無視できるようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{P_1}{A_1} = \cfrac{P_2}{A_2}
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P_2 = \cfrac{A_2}{A_1} P_1
    \end{align}
$$
    
</span>

$P_1$で小さい力を与えることで$P_2$に両者の面積比を掛けた値が圧力として加わり大きなものを動かすことができる．

### 1.2.2. 水平な平面に作用する静水圧

静水圧は，水平な平面に作用するときと鉛直な平面に作用するときで考え方が異なる．

直方体の物体が水に沈んだとき$P_1$と$P_2$が水平な平面に作用する静水圧となる．

このとき，物体には上下からの水圧が加わるが，必ず下からの水圧のほうが大きくなる．

また，$P_1$と$P_2$の力の差を**浮力**と呼び，浮力は必ず上向き方向に作用する．

浮力は水中に入れた物体が押しのけた水の重量分が力として作用しており，このことを**アルキメデスの原理**とよぶ．

浮力は次式により導出される．

<span style="font-size:150%;">
$$
    \begin{align}
        P_1 = P_0 + \rho g h_1 \times d^2
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P_2 = P_0 + \rho g (h_1 + h_2) \times d^2
    \end{align}
$$
    
</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P_z = P_2 - P_1 = \rho g h_2 \times d^2 = \rho g V
    \end{align}
$$
    
</span>

このとき，$P_0$は大気圧$[N / m^2]$，$P_z$は浮力$[N / m^2]$である．

また，物体にかかる重量は次式によって求めることができる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        W = mg = \rho' g V
    \end{align}
$$
    
</span>

このとき，$\rho'$は物体の密度$[N / m^2]$である．

水中に浮かべた物体が沈むかどうかは浮力と物体の重量のつり合いによって決まる．

また，浮力と重量には同じ値（重力加速度$g$，体積$V$）がかけられているため，物体の浮き沈みは水の密度より大きいか小さいかで決まる．

### 1.2.3. 鉛直な平面に作用する静水圧

全水圧や静水圧の式には水深が含まれるが，鉛直な平面に作用するときの水深は水面から図心までの距離となる．

※図心：断面の重心

そのため，鉛直な平面に作用する全水圧と静水圧，全水圧の作用位置は次式によってあらわされる．

下図はある水路を堰がしっかりと止めているときのイメージ図である．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P = \rho g h_G A
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p = \cfrac{P}{A} = \cfrac{\rho g h_G A}{A} = \rho g h_G
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        h_c = h_G + \cfrac{I_G}{h_G A}
    \end{align}
$$
    
</span>

このてき，$h_G$は水面から図心までの距離$[m]$，$h_c$は全水圧の作用位置$[m]$，$I_G$は図心位置における断面２次モーメント$[m^4]$.

よって，全水圧と作用位置は

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P = \rho g h_G A = \rho g \cfrac{H}{2} H B = \cfrac{1}{2} \rho g H^2 B
    \end{align}
$$
    
</span>

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        h_c = h_G + \cfrac{I_G}{h_G A} = \cfrac{H}{2} + \cfrac{B H^3 / 12}{(H / 2) B H} = \cfrac{H}{2} + \cfrac{H}{6} = \cfrac{2}{3} H
    \end{align}
$$
    
</span>

鉛直な平面に作用する全水圧は水面から図心までの距離，全水圧の作用位置には断面２次モーメントが必要となる．

### 1.2.4. 傾斜平面に作用する静水圧

ここでは，傾斜した片面を利用して全水圧と全水圧の作用位置の式の誘導をする．

平面が水面から角度$\theta$だけ傾斜している場合を考える．

全水圧の導出をする

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &dP = \rho g h dA = \rho g y sin \theta dA \\
        &P = \int_A \rho g y sin \theta dA = \rho g sin \theta \int_A y dA
    \end{align}
$$
    
</span>

このとき，$\int ydA$はx軸に関する断面１次モーメントと考えることができ，全水圧の式を次のように変形できる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &\int_A y dA = y_G A \\
        &P = \rho g y_G sin \theta A = \rho g h_G A
    \end{align}
$$
    
</span>

よって，平面に作用する全水圧を誘導することができる．

続いて，全水圧の作用位置$y_c$を求める．

始めに，全水圧のx軸に関するモーメントを積分で考える．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        M = P y_c = \int_A y dP = \int_A \rho g y^2 sin \theta dA = \rho g sin \theta \int_A y^2 dA
    \end{align}
$$
    
</span>

このとき，$\int y^2 dA$はx軸に関する断面２次モーメントと考えることができ，全水圧の作用位置が求まる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &\int_A y^2 dA = I_x \\
        &P y_c = \rho g sin \theta I_x \\
        &y_c = \cfrac{\rho g sin \theta I_x}{P} = \cfrac{\rho g sin \theta I_x}{\rho g y_G sin \theta A} = \cfrac{I_x}{y_G A}
    \end{align}
$$
    
</span>

また，x軸に関する断面２次モーメントは次のように書くことができるため，全水圧の作用位置は次式で定義されている．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &I_x = I_G + y_G^2 A \\
        &y_c = \cfrac{I_G + y_G^2 A}{y_G A} = \cfrac{I_G}{y_G A} + y_G
    \end{align}
$$
    
</span>

全水圧の作用位置は図心より$I_G / (y_G A)$だけ深い位置にあることわかる．

円形の水門に作用する全水圧及び作用位置

まずは，全水圧と全水圧の作用位置を求めるのに必要な物理量を求める．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &y_G = 3 + 0.5 = 3.5[m] \\
        &h_G = 3.5 \times sin 30° = 1.75 [m] \\
        &A = \cfrac{\pi \times 1^2}{2^2} = 0.785 [m^2] \\
        &I_G = \cfrac{\pi \times 1^4}{64} = 0.049 [m^4]\\
    \end{align}
$$
    
</span>

求めた物理量を全水圧及び全水圧の作用位置の式に代入する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &P = \rho g h_G A = 1000 \times 9.8 \times 1.75 \times 0.785 = 13462.75 [N] = 13.46 [kN] \\
        &y_c = \cfrac{I_G}{y_G A} + y_G = \cfrac{0.049}{3.5 \times 0.785} + 3.5 = 3.518 [m] \\
    \end{align}
$$
    
</span>

### 1.2.5. 曲面に作用する静水圧

曲面に作用する静水圧は，静水圧が斜めに作用する．

この静水圧を一つの合力として表すことが困難なため，まずは水平方向と鉛直方向に分解しなければならない．

※奥行幅$B$は$2.0[m]$

上図の円を**ローリングゲート**という．

ローリングゲートはダムの水を堰き止めているゲートの一種である．

ゲートを上下させる際に回転しながら動く．

古いダムによく採用されているが，最近はシェル構造ゲートに変更しつつある．

ローリングゲートに静水圧がかかっている．

圧力は面に対して垂直に作用する．

ローリングゲートの断面は円形なので，静水圧はすべて円の中心に向かって作用する．

ローリングゲートにかかる鉛直方向の力（浮力）をもとめる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P_z &= \rho g V \\
            &= 1000 \times 9.8 \times 2.0 \times \cfrac{\pi \times 3^2}{4} \times \cfrac{1}{2} \\
            &= 69.27 [kN]
    \end{align}
$$
    
</span>

水平方向の力についてもとめる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P_x &= \rho g h_G A \\
            &= 1000 \times 9.8 \times \cfrac{3.0}{2} \times 3.0 \times 2.0 \\
            &= 88.20 [kN]
    \end{align}
$$
    
</span>

次に，鉛直方向の力の作用位置，水平方向の作用位置を計算する．

このとき，鉛直方向の力の作用位置は，水圧が三角形分布で作用することから$2h / 3$となる．

水平方向の力の作用位置は力のモーメントを使って求める．

円の中心は必ずモーメントが０になり，釣り合い式がたてられる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &P_z \times a - P_x \left(\cfrac{3}{2} - \cfrac{2}{3} \right) = 0 \\
        &a = \cfrac{5P_x}{6P_z} = \cfrac{5 \times 88.20}{6 \times 69.27} = 1.061 [m]
    \end{align}
$$
    
</span>

最後に，水平方向の力と鉛直方向の力の合力を求める．

合力の計算は各成分の２乗を足し合わせてから平方根をとれば求まる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P &= \sqrt{(P_z^2 + P_x^2)}  \\
          &= \sqrt{(69.27^2 + 88.20^2)} = 112.15 [kN]
    \end{align}
$$
    
</span>

### 1.2.6. 浮体の安定

**浮体**とは水に浮かんで静止している物体のことであり，この浮体が傾いたときに点灯するのか復元するのか計算する．

浮体には**浮心**（浮力の作用位置）がある．

浮心は浮体が傾くと少しずつずれていく．

そのずれた浮心から鉛直方向に線を伸ばし，浮体の中心線と交わった点を**傾心**または**メタセンター**とよぶ．

また，上図の浮体が転倒することは絶対にない．

それは，傾心が重心より高い位置にあるために半時計回りのモーメントが作用するからである．

そのため，浮心が重心より高い場合は常に安定している．

一方，浮心が重力より低い場合は，転倒し始める．

上図のように，傾心が重心より下にあると転倒モーメントが発生し，傾心が重心より上にあると復元モーメントが発生する．

傾心と重心が同じ位置のときは浮体は釣り合っているため，そのまま静止する．

これを式で表わすと次のようになる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &\overline{MG} = \cfrac{I_y}{V} - \overline{GC} \\
        &\overline{MG} > 0 : 安定 \\
        &\overline{MG} = 0 : 中立 \\
        &\overline{MG} < 0 : 不安定
    \end{align}
$$
    
</span>

このとき，$MG$は傾心高[m], $GC$は浮心から重心までの距離[m]．

上式の誘導をする．

まずは，浮体の傾きを微小とし，浮力によるモーメントから式を立てる．

また，$\int x^2 dA$は$y$軸（奥行き方向の軸）に関する断面２次モーメントと考えることができ，浮心の移動距離が求まる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &P_z e = \int_A \rho g x tan \theta dAx = \rho g tan \theta \int_A x^2 dA \\
        &\int_A x^2 dA = I_y \\
        &P_z e = \rho g Ve = \rho g tan \theta I_y \\
        &e = \cfrac{I_y}{V} tan \theta
    \end{align}
$$
    
</span>

このとき，$e$は浮心の移動距離[m]．

また，浮心の移動距離は三角関数を使った方法でも求めることができる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        e = \overline{MC} tan \theta = (\overline{MG} + \overline{GC}) tan \theta
    \end{align}
$$
    
</span>

この浮心の移動距離の式を組み合わせることにより傾心高の式を導出することができる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &(\overline{MG} + \overline{GC}) tan \theta = \cfrac{I_y}{V} tan \theta \\
        &\overline{MG} = \cfrac{I_y}{V} - \overline{GC}
    \end{align}
$$
    
</span>

下図のようなコンクリートケーソンがある．

そこに砂を入れて海水に浮かべた時の喫水$H$（コンクリートが海水に浸かっている高さ），重心位置，浮心位置を求め，ケーソンの安定性を評価せよ．

ケーソンとは，防波堤や岸壁を作る際に使われる鉄筋コンクリートでできた箱

ただし，コンクリートの単位重量は$23.52[kN / m^3]$，海水の単位重量は$1.025[kN/m^3]$とする．

まずは，コンクリートの体積と重量を求める．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        V &= 2.0 \times 2.0 \times 3.0 - 1.8 \times 1.9 \times 2.8 \\
          &= 2.424[m^3] \\
        W &= 2.424 \times 23.52 = 57.01[kN]
    \end{align}
$$
    
</span>

次に，浮力を求める．

このとき，喫水を$H$として計算する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        P_z &= 1.025 \times 9.8 \times 2.0 \times 3.0 \times H \\
            &= 60.27H [kN]
    \end{align}
$$
    
</span>

また，浮力と重量は釣り合っているため，喫水$H$を計算することができる．

また，浮心から底面Bまでの距離は，喫水の半分なので$H / 2$で求められる．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &H = \cfrac{57.01}{60.27} = 0.946[m] \\
        &\overline{CB} = \cfrac{H}{2} = \cfrac{0.946}{2} = 0.473[m]
    \end{align}
$$
    
</span>

上下非対称の物体の場合，重心が物体の高さの半分にならないので，底面に関するモーメントを考え，重心から底面までの距離を求める．

その結果を用いて重心から浮心までの距離を計算する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &W \times \overline{GB} \\
        &= 23.52 \times \left(2.0 \times 2.0 \times 3.0 \times \cfrac{2.0}{2} - 1.8 \times 1.9 \times 2.8 \times \left(0.1 + \cfrac{1.9}{2} \right) \right) \\
        &= 45.75[kN \cdot m] \\
        &\overline{GB} = \cfrac{45.75}{W} = \cfrac{45.75}{57.01} = 0.802[m] \\
        &\overline{GC} = \overline{GB} - \overline{CB} = 0.802 - 0.473
= 0.329 [m]    
    \end{align}
$$
    
</span>

y軸（底面）における断面２次モーメントを求める．

このとき，y軸のとり方によって断面２次モーメントが２パターン存在する．

浮体の安定を計算するときは安全性を考慮し，小さい値を採用する．

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        &I_{y1} = \cfrac{b h^3}{12} = \cfrac{2.0 \times 3.0^3}{12} = 4.5[m^4] \\
        &I_{y2} = \cfrac{b^3 h}{12} = \cfrac{2.0^3 \times 3.0}{12} = 2.0[m^4]
    \end{align}
$$
    
</span>

最後に，傾心高を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &\overline{MG} = \cfrac{I_y}{V} - \overline{GC} = \cfrac{2.0}{2.424} - 0.329 = 0.496[m]
    \end{align}
$$
    
</span>

コンクリートケーソンは安定であることがわかる．

浮体の安定性は傾心高を求めることに照査することができる．

### 1.2.7. 相対的な静水圧

水の入ったバケツを荷台に乗せて一定の速度で動かすと，バケツ内の水は慣性力を受け，水面が傾き始める．

そのうち傾きはあるところで一定を保ち続けることになる．

このような状態を**相対的静止**という．

荷台を加速させると，水には常に慣性力が作用していることになる．

この慣性力と水の重量から角度を計算することができる．

<span style="font-size:150%;">
$$
    \begin{align}
        tan \theta = \cfrac{F}{W} = \cfrac{ma}{mg} = \cfrac{a}{g}
    \end{align}
$$

</span>

次に，地面が傾いている場合を考える．

地面の傾きの分だけ慣性力も傾く．

そのため，慣性力を水平方向と鉛直方向に分解する必要がある．

<span style="font-size:150%;">
$$
    \begin{align}
        tan \theta = \cfrac{F_x}{W + F_z} = \cfrac{ma cos \varphi}{mg + ma sin\varphi} = \cfrac{a cos\varphi}{g + a sin\varphi}
    \end{align}
$$

</span>

さらに，水が回転運動する場合を考える．

バケツの中に水を入れ，水を回転させると中心付近の水面が低くなり，端の水面は高くなる．

これは，水に遠心力が加わるためである．

<span style="font-size:150%;">
$$
    \begin{align}
        tan \theta = \cfrac{F}{W} = \cfrac{m \omega^2 x}{mg} = \cfrac{\omega^2 x}{g}
    \end{align}
$$

</span>

このとき，$\omega$は角速度$[rad / s]$.

また，水面の傾きは$dz / dz$と表わすことができる．

この式を積分することで，水面高さの式を誘導することができる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{dz}{dx} = \cfrac{\omega^2}{g} x \\
        &z = \cfrac{\omega^2}{2g} x^2 + C
    \end{align}
$$

</span>

上式に初期条件$(x = 0, z = h_0)$を当てはめることで，積分定数および水面高さが求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        &C = h_0 \\
        &z = \cfrac{\omega^2}{2g} x^2 + h_0
    \end{align}
$$

</span>

直径20cmの円筒容器内の水を毎分30回転させると，端の水面は中心の水面より何cm高くなるか．

１秒間の回転数と角速度の関係は次式で表わされる．

<span style="font-size:150%;">
$$
    \begin{align}
        &n = \cfrac{\omega}{2 \pi} \\
        &\omega = 2 \pi n = 2 \pi \times \cfrac{30}{60} = \pi [rad / s]
    \end{align}
$$

</span>

角速度が求まったので，中心の水面から端の水面までの高さは次のように計算できる．

<span style="font-size:150%;">
$$
    \begin{align}
        &z - h_0 = \cfrac{\omega^2}{2g} x^2 = \cfrac{\pi^2}{2 \times 9.8} 0.1^2 = 0.005[m] = 0.5[cm]
    \end{align}
$$

</span>

相対的な静水状態は直進運動と回転運動の２つに分けられる．

直進運動の傾きは慣性力/重量で，回転運動の傾きは遠心力/重量で求めることができる．

## 1.3. 流れの基礎理論

### 1.3.1 流体の分類

流体とは気体と液体の総称であり，圧縮性流体と非圧縮性流体に分類することができる．

また，粘性を考慮した流体を粘性流体，粘性を無視した物体を非粘性流体という．

非圧縮性かつ非粘性の流体は**理想流体**または**完全流体**と呼ばれ，運動法則を簡単にすることができる．

水は一般的に非圧縮性の粘性流体として扱われる．

水は一般的に連続した通路の中，すなわち水路内を流れている．

また，水路は水面の有無によって**開水路**と**菅水路**に分けることができる．

菅水路の例としては河川や用水路，菅水路の例としては上水管が挙げられる．

ちなみに，下水管は管内の水が一杯にならないため，開水路として扱われることが多い．

これら水路で流れる水の速さを**流速**，水の流れる断面積を**流積**という．

さらに，この２つを掛け合わせたものを**流量**といい，流量は水が単位時間当たりに流れる量を表している．

また，水路の壁に接している部分の長さを**潤辺**，流積を潤辺で割った物理量を径深という．

開水路，菅水路における潤辺は下図のようなイメージであり，径深を式で表わすと次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        R = \cfrac{A}{S}
    \end{align}
$$

</span>

このとき，$R$は径深[m]，$A$は流積$[m^2]$，$S$は潤辺$[m]$.

上図の左側は開水路の流れを表しており，右側は菅水路の流れを表している。

径深は図で表わすことができず，非常にイメージがつかみにくい．

なので，歴史を追って説明する．

昔の人は水深の大きさから流速を計算していたが，そこの丸い水路に悩まされた．

なぜなら，底が丸い水路は水深をどこに設定すればよいのかわからなかったからである．

そこで，流積を何かの長さで割り，換算水深を出すことにした．

候補は水面長と潤辺の二つあった．

流速は粘性（摩擦）に大きく左右される．

そのために，平均水深も摩擦の情報を持っている必要があり，自然と流積を割る長さは潤辺に決まった．

したがって，径深は水路側壁の摩擦力を考慮した換算水深といえる．

流速の式としては**マニングの式**がある．

マニングの式では，径深は2 / 3乗されており，径深が大きくなれば流速は小さくなる．

<span style="font-size:150%;">
$$
    \begin{align}
        v = \cfrac{1}{n} R^{2 / 3} I^{1 / 2}
    \end{align}
$$

</span>

このとき，$v$は流速[m / s]，$n$は粗度係数$[m^{-1/3}/s]$，$I$は動水勾配[単位なし]．

水路での物理量として，流速，流積，流量，潤辺，径深の５つのパラメータがある．

### 1.3.2 流れの分類

開水路の流れは，**定常流**と**非定常流**にわけることができる．

定常流は時間によって流量が変化しない流れであり，平常時の河川や水路などが当てはまる．

一方，非定常流は時間によって流量（流積や流速)が変化する流れであり，洪水時の河川や感潮域（潮の満ち引きによって水深が変わる領域）などが当てはまる．

さらに，定常流は**等流**と**不等流**に分けられる．

等流は空間が変化しない流れ，すなわち流積や流速が常に一定の流れである．

一方，不等流は流量は一定なのだが，流積と流速と流速が場所によって変化する流れである．

開水路の流れは長波の伝播速度（波速でもよい）でも分類することができる．

波の大きさは波長と波高によって決まるが，長波は波長が水深の２５倍以上大きいときの波をいう．

この長波の波速と流速の比によって開水路の流れが分類されている．

式にすると次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        F_r = \cfrac{v}{c} = \cfrac{v}{\sqrt{gh}}
    \end{align}
$$

</span>

このとき，$F_r$はフルード数[単位なし]，$c$は波速[m/s].

フルード数が１より小さい流れを**常流**，フルード数が１を超える流れを**射流**という．

また，フルード数が１の流れを**限界流**という．

常流時の川に大きな石を投げ入れると，石が落ちたところから水面の全方向に波が広がる．

一方，射流の時の川に大きな石を投げ入れても波が立つことはない．

これは，川の流速が波速よりも速いからである．

台所の蛇口を勢いよくひねると下の写真のようになる．

!

流れが二つに分かれているのがはっきり見える．

このとき，内側が射流で外側が常流である．

射流から常流に移っている部分の水深が少し大きくなっている．

この現象を**跳水**という．

菅水路の流れの分類は粒子の動きに注目する．

管にインクを垂れ流しているとすると次のようになる．

流速が遅いときインクはまっすぐに流れ続ける．

一方で，流速が速くなると，インクは途中で乱れてしまう．

ゆっくり流れているときは，粘性力によって粒子は自由に動くことができない．

そのため，均一な状態で流れていく．

しかし，速度が上昇すると粘性力より慣性力が強くなる．

その結果，粒子が自由に動き回ってしまい，インクが乱れる．

粒子が層になる流れを**層流**とよび，粒子が乱れた流れを**乱流**とよぶ．

また，次式によって流れを調べることができる．

<span style="font-size:150%;">
$$
    \begin{align}
        R_e = \cfrac{vD}{\nu}
    \end{align}
$$

</span>

このとき，$R_e$はレイノルズ数[単位なし]，$D$は管径[m]，$\nu$は動粘性係数or動粘度$[m^2/s]$．

**レイノルズ数**の値は2000より小さければ層流，2000～4000は過渡状態，4000より大きければ乱流となる．

過渡状態→流れがある状態から別の状態へ移行する過程の流れ

そのため，平均流速が大きく，動粘性係数が小さければより乱流になりやすくなる．

また，管径が大きくなればなるほど壁からの距離が粘性力は小さくなる．

その結果，乱流になりやすくなる．

開水路でレイノルズ数を用いたい場合は管径の代わりに径深を使用する，

### 1.3.3. 相似則

水理学の分野では，河川や海岸の解析をするときに模型実験がよく行われる．

模型実験とは，現地の状況を縮小した模型を作製し，模型による測定値から原型の真値を求める実験のことである．

このとき，原型と模型の寸法の比がすべて等しい相似を**幾何学的相似**といい，流速や粘性力などの作用する力の比がすべて等しい相似を**力学的相似**という．

水力学で力学的相似を行うとき，流速や流量，水圧，粘性力など作用する力の比をすべて等しくしようとすると多大な労力と時間が必要になる．

そこで，開水路の模型ではフルード数，菅水路の模型ではレイノルズ数がよく用いられる．

**フルードの相似則**は重力が影響しているときに慣性力と重量の比を一定にしたものであり，**レイノルズの相似則**は粘性力が作用しているときに慣性力と粘性力の比を一定にしたものとなる．

<span style="font-size:150%;">
$$
    \begin{align}
        F_r = \sqrt{\cfrac{ma}{mg}} = \sqrt{\cfrac{\rho L^3 L T^{-2}}{\rho L^3 g}} = \sqrt{\cfrac{\rho L^2 L^2 T^{-2}}{\rho L^3 g}} = \sqrt{\cfrac{\rho L^2 V^2}{\rho L^3 g}} = \sqrt{\cfrac{V^2}{L g}} = \cfrac{V}{\sqrt{L g}}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        R_e = \cfrac{ma}{\tau A} = \cfrac{\rho L^2 V^2}{\mu (du/dz) A} = \cfrac{\rho L^2 V^2}{\mu (V / L) L^2} = \cfrac{\rho L V}{\mu} = \cfrac{VL}{\nu}
    \end{align}
$$

</span>

このとき，$L$は代表長さ[m]，$V$は代表速度[m/s]．

縮尺1/100のダムの模型を作製し，越流の流速を測定したところ0.5[m/s]であった．では原型の流速は．

ダムの越流は壁に触れておらず粘性力がほとんど働いていないため，重力の影響が大きいと仮定できる．

越流　水があふれ出ること．

そのため，フルード数を使って流速を求める．

模型の添え字は$m$，原型の添え字は$p$を使用し，比により流速を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        F_r = \cfrac{v_p}{\sqrt{gh_p}} = \cfrac{v_m}{\sqrt{gh_m}}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        v_p = \cfrac{v_m}{\sqrt{gh_m}} \times \sqrt{gh_p} = \sqrt{\cfrac{h_p}{h_m}} v_m = \sqrt{100} \times 0.5 = 5.0[m/s]
    \end{align}
$$

</span>

縮尺1/25の河川の模型を作製し，原型の流量を測定したところ，60$[m^3/s]$であった．模型の流量は．

河川の場合もフルード数を使って求める．

<span style="font-size:150%;">
$$
    \begin{align}
        F_r = \cfrac{v_p}{\sqrt{gh_p}} = \cfrac{v_m}{\sqrt{gh_m}}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_p}{v_m} = \cfrac{\sqrt{gh_m}}{\sqrt{gh_p}} = \sqrt{\cfrac{h_m}{h_p}} = \sqrt{\cfrac{1}{25}} = \cfrac{1}{5}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{Q_p}{Q_m} = \cfrac{A_p v_p}{A_m v_m} = \left( \cfrac{1}{25} \right)^2 \times \cfrac{1}{5} = \cfrac{1}{3125}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        Q_p = \cfrac{1}{3125} \times 60 = 0.0192[m^3 / s]
    \end{align}
$$

</span>

相似則として使える無次元量は，フルード数，レイノルズ数のほかにウェーバー数というのが存在する．

ウェーバー数は慣性力と表面張力の比で表わされた無次元量．

<span style="font-size:150%;">
$$
    \begin{align}
        W_e = \cfrac{\rho L^2 V^2}{T}
    \end{align}
$$

</span>

### 1.3.4. 流れの連続性

下図のような菅水路は壁で区切られているため，水が出入りすることがない．

そのため，**質量保存の法則**から管内の質量は常に一定であることがうかがえる．

<span style="font-size:150%;">
$$
    \begin{align}
        \rho_1 A_1 v_1 dt = \rho_2 A_2 v_2 dt = const.
    \end{align}
$$

</span>

const.はconstantの略であり，一定を意味する．

水は非圧縮性流体であり，温度などを考慮しなければ密度は一定となる．

<span style="font-size:150%;">
$$
    \begin{align}
        A_1 v_1 = A_2 v_2 = const.
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        Q = Av = const.
    \end{align}
$$

</span>

この関係式は**連続の式**と呼ばれ，水力学の三大公式の一つである．

管路における断面②の平均流速をもとめる．ただし断面①の管径は0.2[m]，流速は0.5[m/s]，断面②の管径は0.1[m]とする．

連続の式から断面②の平均流速を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        v_2 = \cfrac{A_1}{A_2} = \cfrac{\pi d_1^2 / 4}{\pi d_2^2 / 4} v_1 = \cfrac{d_1^2}{d_2^2} v_1 = \cfrac{0.2^2}{0.1^2} \times 0.5 = 2.0 m/s
    \end{align}
$$

</span>

定常流れにおける流量は一定であり，流量に関する式を連続の式という．

### 1.3.5. ベルヌーイの定理

**ベルヌーイの定理**は1738年にスイスの物理学者である**ダニエル・ベルヌーイ**が発見した運動方程式であり，流体の速度が増加すると圧力が下がることを証明した．

連続の式は質量保存則の法則から求めたが，ベルヌーイの定理は**エネルギー保存の法則**から求める．

流体が持つ全エネルギーは，運動エネルギーと位置エネルギーの総和となる．

また，流体は圧力の形でもエネルギーを蓄えることができ，静水圧の式を変形することで圧力によるエネルギーを求められる．

<span style="font-size:150%;">
$$
    \begin{align}
        E = \cfrac{mv^2}{2} + mgz + mg \cfrac{p}{\rho g}
    \end{align}
$$

</span>

このとき，$E$はエネルギー[J]，$z$は基準線からの高さ[m]．

水力学では，運動エネルギーのことを**動圧**，圧力によるエネルギーのことを**静圧**とよび，動圧＋静圧を**全圧**という．

また，静圧と静水圧は異なるものである．

静圧は圧力によって生じるエネルギー量を表しており，単位は[J]または[N・m]を使用する．

一方，静水圧は水による圧力を表しており，単位は[Pa]または$[N/m^2]$を使用する．

では，エネルギー保存の法則から式を立てる．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{mv_1^2}{2} + mgz_1 + mg \cfrac{p_1}{\rho g} = \cfrac{mv_2^2}{2} + mgz_2 + mg \cfrac{p_2}{\rho g}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g}
    \end{align}
$$

</span>

この式はエネルギーを単位重量で割った式であるので，単位換算を行っていくと単位は長さ[m]になる．

この長さの単位は水深ではなく水頭を表す．

$v^2/(2g)$は速度水頭，$p/(\rho g)$は圧力水頭，$z$は位置水頭とよぶ．

上式をさらに端的にすると次式のようになり，この式を**ベルヌーイの定理**という．

円形断面のパイプにおける圧力$p_2$を求める．

ただし，$p_1 = 30[kPa]$，$v_1 = 1.0[m/s]$，$d_1 = 0.2[m]$，$d_2 = 0.1[m]$とする．

まずは，連続の式を使って，速度$v_2$を計算する．

<span style="font-size:150%;">
$$
    \begin{align}
        v_2 = \cfrac{A_1}{A_2} v_1 = \cfrac{\pi d_1^2 / 4}{\pi d_2^2 / 4} v_1 = \cfrac{d_1^2}{d_2^2} v_1 = \cfrac{0.2^2}{0.1^2} \times 1.0 = 4.0 m /s
    \end{align}
$$

</span>

次に，ベルヌーイの式から圧力$p_2$を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{1.0^2}{2 \times 9.8} + 5.0 + \cfrac{30 \times 1000}{1000 \times 9.8} = \cfrac{4.0^2}{2 \times 9.8} + 2.0 + \cfrac{p_2}{1000 \times 9.8}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{1.0^2}{2 \times 9.8} + 5.0 + \cfrac{30 \times 1000}{1000 \times 9.8} = \cfrac{4.0^2}{2 \times 9.8} + 2.0 + \cfrac{p_2}{1000 \times 9.8}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        p_2 = 51900 Pa = 51.9[kPa]
    \end{align}
$$

</span>

下図のようにとても大きい水槽に水を入れた後，側壁の下部に小孔を開け水を流出させた．このとき，小孔から流出した水の速度を求めよ．

まずは，水槽の水面と小孔直後の２点でベルヌーイの定理を適応させる．

このとき，小孔を基準線として考えると，$z_1 = H$，$z_2 = 0$となる．

また，両点とも大気圧に接してるので，$p_1 = 0$，$p_2 = 0$となる．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_1^2}{2g} + H + 0 = \cfrac{v_2^2}{2g} + 0 + 0
    \end{align}
$$

</span>

また，水面の面積は小孔断面積に比べて非常に大きく，水面の効果速度はかなり遅くなる．

このとき，速度$v_1 \fallingdotseq 0$とみなすことができ，その結果，速度$v_2$を求めることができる．

<span style="font-size:150%;">
$$
    \begin{align}
        v_2 = \sqrt{2gH}
    \end{align}
$$

</span>

小孔から噴出する水の速度は，水深の平方根に比例することがわかる．

また，この関係を**トリチェリの定理**といい，イタリアの物理学者**エヴァンジェリスタ・トリチェリ**によって1644年に発見される．

翼の上側と下側で速度差が生じている．

上側は速度が速く，下側は速度が遅くなっている．

つまり，上側は下側より圧力が小さくなっている．

その結果，下から上に向かって新しい力が作用し始める．

この力を**揚力**とよぶ．

速度差が生まれる本当の理由は流体の循環にある．

翼の後端のとがった形状が非常に重要である．

この尖った形状により翼の上面の気流，下面の気流が後端でスムーズに合流できるようになる．

このように翼の角度や形状の工夫することで，翼の上下を通る流体を滑らかに合流させるように促すことをクッタ条件を満たすという

クッタ条件を満たすと翼の表面を取り巻くように気流が循環しようとする力が発生する．

この図をよく見ると，翼の上側は流体の流れを後押ししており，下側は流体の流れに抵抗するような力が作用している．

その結果，速度差や圧力差すなわち揚力が生じている．

### 1.3.6. ベルヌーイの定理の応用

ベルヌーイの定理を応用して流速を測定する装置を**ピトー管**，管水路の流量を測定する装置を**ベンチュリメーター**，開水路の流量を測定する装置を**ベンチュリフルーム**という．

ピトー管はL字型の細い管でできており，ピトー管の先端を測定場所の少し後ろに置くと流速を求めることができる．

流れの中にピトー管を置くと管入り口に流速が０になる点ができ，これを**よどみ点**とよぶ．

速度が速くなると圧力は低くなるので，よどみ点では圧力が正確に測定でき，この圧力から流速が算出できる．

このとき，基準線の高さは同じなので$z_1 - z_2 = 0$となり，よどみ点から$v_2 = 0$となる．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{v_1^2}{2g} + \cfrac{p_1}{\rho g} = 0 + \cfrac{p_2}{\rho g}
    \end{align}
$$

</span>

このとき，２点間の圧力水頭の差を$h$と置き換え，速度$v_1$を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        h = \cfrac{p_2}{\rho g} - \cfrac{p_1}{\rho g}
    \end{align}
$$

</span>

<span style="font-size:150%;">
$$
    \begin{align}
        v_1 = \sqrt{2gh}
    \end{align}
$$

</span>

次にベンチュリメーターである．

ベンチュリメーターは管水路に断面収縮部を設けており，そのときの圧力差を利用して流量を求める装置になる．

ベルヌーイの定理から導出をしていく．

ベンチュリメーターもピトー管と同様に$z_1 - z_2 = 0$になる．

また，２点間の圧力水頭の差を$h$と置き換え，式変形する．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} \\
        &\cfrac{v_1^2}{2g} + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} +  \cfrac{p_2}{\rho g} \\
        &\cfrac{v_2^2}{2g} - \cfrac{v_1^2}{2g} = \cfrac{p_1}{\rho g} - \cfrac{p_2}{\rho g} = h \\
        &v_2^2 - v_1^2 = 2gh
    \end{align}
$$

</span>

次に，連続の式を使って速度から流量に変換する．

すると，ベンチュリメーターの式の誘導をする．

<span style="font-size:150%;">
$$
    \begin{align}
        &Q = A_1 v_1 = A_2 v_2 \\
        &\left(\cfrac{Q}{A_2} \right)^2 - \left(\cfrac{Q}{A_1} \right)^2 = 2gh \\
        &\cfrac{A_1^2 - A_2^2}{A_1^2 A_2^2} Q^2 = 2gh \\
        &Q^2 = \cfrac{A_1^2 A_2^2}{A_1^2 - A_2^2} 2gh \\
        &Q = \cfrac{A_1 A_2}{\sqrt{A_1^2 - A_2^2}} \sqrt{2gh}
    \end{align}
$$

</span>

最後にベンチュリフルーム．

ベンチュリメーターは管の途中に断面収縮部に対し，ベンチュリフルームは開水路の一部に狭い部分を作ることで流量を大きくし，水位を下げる．

この水位の低下量を測定することで流量を求める装置である．

図のように幅を狭くしたため水深が変化する．

このとき，断面内で流速が一様で水平になっていると考えると，速度が次のように求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        &v_1 = \cfrac{Q}{b_1 h_1} \\
        &v_2 = \cfrac{Q}{b_2 h_2}
    \end{align}
$$

</span>

これら速度の式をベルヌーイの定理に代入することで流量が求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} \\
        &\cfrac{1}{2g} \left(\cfrac{Q}{b_1 h_1} \right)^2 + h_1 = \cfrac{1}{2g} \left(\cfrac{Q}{b_2 h_2} \right)^2 + h_2  \\
        &Q^2 \left\{ \cfrac{1}{(b_1 h_1)^2} - \cfrac{1}{(b_2 h_2)^2} \right\} = 2g(h_2 - h_1)\\
        &Q = \sqrt{\cfrac{2g(h_2 - h_1)}{1 / (b_1 h_1)^2 - 1 / (b_2 h_2)^2}}
    \end{align}
$$

</span>

### 1.3.7. 運動量方程式

定常流における**運動量方程式**を誘導する．

まずは，ニュートンの運動方程式から式変形をしていく．

<span style="font-size:150%;">
$$
    \begin{align}
        \mathbf{F} = m \mathbf{a} = m \cfrac{d \mathbf{v}}{dt} 
    \end{align}
$$

</span>

細字で書かれたアルファベットはスカラー量，太字で書かれたアルファベットはベクトル量を表す．

スカラー量は物理量の大きさだけを表現しており，質量や時間，温度が例としてあげられる．

一方，ベクトル量は物理量の大きさと方向を持った物理量であり，力，速度，加速度などがある．

ベクトル量は方向の情報も持っているため，負の値も存在する．

上式の$md \mathbf{v}$は運動量といい，物体の運動の激しさを示す量である．

また，$d \mathbf{v}$や$dt$は微小範囲における変化量を示している．

いま，２個の質点を仮定すると，さらに式変形が行える．

<span style="font-size:150%;">
$$
    \begin{align}
        \mathbf{F} = \cfrac{m (\mathbf{v_2} - \mathbf{v_1})}{t_2 - t_1} = \rho Q (\mathbf{v_2} - \mathbf{v_1}) 
    \end{align}
$$

</span>

この式を運動量方程式といい，運動量の変化を表す．

また，外力が作用していないときは**運動量保存の法則**が成り立つ．

一般的に，運動量方程式を適応させる境界面を検査面といい，この検査面に囲まれた領域を**検査領域**という．

下図のような内径からある流速で水が噴出されたとき，壁面はどのくらいの力を受けているか．

図の破線は検査領域を示している．

検査領域の最も左を断面①，検査領域の最も右を断面②として運動量方程式を立てる．

断面①の力は大気圧しか作用していないので０になる．

また，断面②の流速は壁にぶつかっているので０になる．

流量$Q$を連続の式で求め，式に代入すると外力が算出できる．

<span style="font-size:150%;">
$$
    \begin{align}
        &0 - F = \rho Q (0 - v) \\
        &F = \rho \cfrac{\pi D^2}{4} v^2
    \end{align}
$$

</span>

下図のように水の噴出が角度30°の平板の衝突するときの反力を求める．

また，平板上で摩擦抵抗を考慮しないときの流量$Q_2$，$Q_3$を求めよ．

ただし，流量$Q_1 = 0.2[m^3]$，$v_1 = 50[m/s]$とする．

壁面に直角なx軸方向と，壁面に沿うy軸方向に分ける．

まずは，x軸方向の運動量方程式を立て，外力を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &0 - F = 0 - \rho Q_1 v_1 sin 30° \\
        &F = 1000 \times 0.2 \times 50 \times \cfrac{1}{2} = 5000N = 5.0[kN]
    \end{align}
$$

</span>

次に，y軸方向の運動量方程式をたてる．

このとき，摩擦抵抗を考慮しないのであれば速度は$v_1 = v_2 = v_3$となる．

<span style="font-size:150%;">
$$
    \begin{align}
        &0 = \rho Q_2 v_2 + \rho Q_1 v_1 cos 30° - \rho Q_3 v_3 \\
        &0 = \rho Q_2 v + \rho Q_1 v cos 30° - \rho Q_3 v \\
        &0 = Q_2 v + Q_1 cos 30° - Q_3  \\
    また，\\
        &Q_1 = Q_2 + Q_3
    \end{align}
$$

</span>

この連立方程式から$Q_2$を消去することにより流量を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &Q_1 (1 + cos30°) = 2 Q_3 \\
        &Q_3 = \cfrac{1}{2} (1 + cos30°) \times 0.2 = 0.187[m^3/s] \\
        &Q_2 = Q_1 - Q_3 = 0.2 - 0.187 = 0.013[m^3/s]
    \end{align}
$$

</span>

下図のように堰を超える流れにおいて，単位幅流量が$q = 1.7[m^3/s]$のとき，堰幅$1[m]$あたりにかかる力を求める．

単位幅流量とは幅1[m]あたりの流量という．

まずは，全水圧と速度を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &P_1 = \rho g h_G A = \rho g \cfrac{h_1}{2} h_1 = \cfrac{1000 \times 9.8 \times 3.0^2}{2} = 44.1[kN] \\
        &P_2 = \rho g h_G A = \rho g \cfrac{h_2}{2} h_2 = \cfrac{1000 \times 9.8 \times 1.0^2}{2} = 4.9[kN]\\
        &v_1 = \cfrac{q}{A_1} = \cfrac{q}{h_1} = \cfrac{1.7}{3.0} = 0.57[m/s] \\
        &v_2 = \cfrac{q}{A_2} = \cfrac{q}{h_2} = \cfrac{1.7}{1.0} = 1.70[m/s]
    \end{align}
$$

</span>

次に運動量方程式から外力を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &P_1 - P_2 - F = \rho q (v_2 - v_1) \\
        &F = P_1 - P_2 - \rho q (v_2 - v_1) = 44.1 - 4.9 - 1.0 \times 9.8(1.70 - 0.57) = 37.3[kN]\\
    \end{align}
$$

</span>

## 1.4. オリフィス，水門，堰

### 1.4.1. オリフィス

流量を制御するために水槽の底や側面に設けた小孔を**オリフィス**という．

オリフィスには，小型オリフィス，長方形大型オリフィス，円形大型オリフィス，三角形大型オリフィス，潜りオリフィスと様々な種類がある．

オリフィスによる流量の式の誘導を行う．

#### ①小型オリフィス

小型オリフィスの流出速度はトリチェスの定理に従う．

しかし，トリチェスの定理は理論値であるため，実測値に近づけるために損失水頭を考える．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + f \cfrac{v_2^2}{2g} \\
        &0 + h + 0 = \cfrac{v_2^2}{2g} + 0 + 0 + \cfrac{v_2^2}{2g} \\
        &v_2^2(1 + f) = 2gh \\
        &v_2 = \cfrac{1}{\sqrt{1 + f}} \sqrt{2gh}\\
        &v_2 = C_v \sqrt{2gh}
    \end{align}
$$

</span>

このとき，$f$は摩擦損失係数[単位なし]，$C_v$は流速係数[単位なし]．

**摩擦損失係数**は摩擦によるエネルギーの損失を表した係数であり，**流速係数**は理論値を実測値に近づける係数である．

また，流速係数は一般的に0.95～0.99の値をとる．

次に連続の式を使って流量を導出する．

小型オリフィスを出たあとの水はそのまま拡散していくイメージがあるが，流出した直後の流積は一度収縮する．

この収縮した流積を**ベナコントラクタ**とよぶ．

さらに，このベナコントラクタの断面とオリフィスの断面の比を**収縮係数**という．

ちなみに，収縮係数の値は0.6～0.7程度となる．

<span style="font-size:150%;">
$$
    \begin{align}
        &C_c = \cfrac{a_0}{a} \\
        &Q = Av = a_0 v = C_c C_v a \sqrt{2gh} = Ca \sqrt{2gh}
    \end{align}
$$

</span>

このとき，$C_c$は収縮係数[単位なし]，$C$は流量係数[単位なし]．

**流量係数**は収縮係数と流速係数をかければ求めることができ，0.6程度の値をとる．

#### ②長方形大型オリフィス

長方形大型オリフィスの上縁と下縁は流速差が大きくなっていくため，流速を一様と考えることができない．

また，長方形大型オリフィスであるため水面の下降速度も無視できないとすると，ベルヌーイの定理を立て直す必要がある．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + f \cfrac{v_2^2}{2g} \\
        &\cfrac{v_1^2}{2g} + z + 0 = \cfrac{v_2^2}{2g} + 0 + 0 + \cfrac{v_2^2}{2g} \\
    \end{align}
$$

</span>

水面の速度水頭を$h_a$と置き換え，式変形を続ける．

<span style="font-size:150%;">
$$
    \begin{align}
        &h_a = \cfrac{v_1^2}{2g} = \cfrac{v_a^2}{2g} \\
        &v_2^2 (1 + f) = 2g (z + h_a) \\
        &v_2 = \cfrac{1}{\sqrt{1 + f}} \sqrt{2g(z + h_a)} \\
        &v_2 = C_v \sqrt{2g(z + h_a)}
    \end{align}
$$

</span>

このとき，$v_a$は接近流速[m/s]，$h_a$は接近流速水頭[m]．

水面の降下速度を接近流速，水面の速度水頭を接近流速水頭と呼ぶ．

次に流量を求めるが，連続式を使って深さ方向に積分する必要がある．

<span style="font-size:150%;">
$$
    \begin{align}
        Q &= C_v C_c b_0 \sqrt{2g} \int_{h_1}^{h_2}\sqrt{z + h_a}dz \\
          &= C b_0 \sqrt{2g} \left[\cfrac{2}{3} (z + h_a)^{3 / 2} \right]_{h_1}^{h_2} \\
          &= \cfrac{2}{3} C b_0 \sqrt{2g} \{(h_2 + h_a)^{3 / 2} - (h_1 + h_a)^{3 / 2}\} 
    \end{align}
$$

</span>

さらに，接近流速水頭を考慮しない場合を考える．すると，

<span style="font-size:150%;">
$$
    \begin{align}
        Q = \cfrac{2}{3} C b_0 \sqrt{2g} \{(h_2)^{3 / 2} - (h_1)^{3 / 2}\} 
    \end{align}
$$

</span>

このとき，水深$h_1$と$h_2$のちょうど半分の位置を$h$とする．

そのときの水深$h_1$，$h_2$の式を上式に代入する．

<span style="font-size:150%;">
$$
    \begin{align}
        h_1 &= h - \cfrac{h_0}{2} \\
        h_2 &= h + \cfrac{h_0}{2} \\
        Q &= \cfrac{2}{3} C b_0 \sqrt{2g} \left\{ \left( h + \cfrac{h_0}{2} \right)^{3/2} - \left( h - \cfrac{h_0}{2} \right)^{3/2}\right\} \\
          &= \cfrac{2}{3} C b_0 \sqrt{2gh} \cdot h \left\{ \left( 1 + \cfrac{h_0}{2h} \right)^{3/2} - \left( 1 - \cfrac{h_0}{2h} \right)^{3/2}\right\}
    \end{align}
$$
</span>

級数展開（テイラー展開）をすると，さらに式変形することができる．

$x = \cfrac{h_0}{h}$ とすると

<span style="font-size:150%;">
$$
    \begin{align}
        \left( 1 + \cfrac{h_0}{2h} \right)^{3/2} = \left( 1 + \cfrac{1}{2} x \right)^{3/2}
    \end{align}
$$
</span>

$f(x) = \left( 1 + \cfrac{1}{2} x \right)^{3/2}$として，マクロリーン展開すると

<span style="font-size:150%;">
$$
    \begin{align}
        &f(x) = 1 + \cfrac{3}{4} x + \cfrac{3}{32} x^2 -\cfrac{1}{128} x^3 + \cfrac{3}{2048} x^4 - \cfrac{3}{8192} x^5 + ・・・
    \end{align}
$$
</span>

同様に計算し，まとめると

<span style="font-size:150%;">
$$
    \begin{align}
        Q &= C \left(\cfrac{1}{96} \left(\cfrac{h_0}{h} \right)^2 - \cfrac{1}{2048} \left(\cfrac{h_0}{h} \right)^2 - ... \right) b_0 h_0 \sqrt{2gh} \\
          &= C' b_0 h_0 \sqrt{2gh}
    \end{align}
$$
</span>

#### ③円形大型オリフィス

円形大型オリフィスは真円でできていると仮定し，円の中心位置までの深さを水深$h$と置く．

まずは，円の幅，深さ，微小深さの式を三角関数を使って求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &b_0 = 2rsin \theta \\
        &z = h - r cos \theta \\
        &dz = rsin \theta d \theta
    \end{align}
$$
</span>

これらの式を連続の式に代入し，積分する．

また，計算を簡単にするため，接近流速を考えないとする．

<span style="font-size:150%;">
$$
    \begin{align}
        Q &= C_v C_c \sqrt{2g} \int_{h - r}^{h + r} b_0 \sqrt{z} dz \\
          &= 2 C r^2 \sqrt{2gh} \int_0^{\pi} sin^2 \theta \left( 1 - \cfrac{r}{h} cos \theta \right)^{1 / 2} d \theta
    \end{align}
$$
</span>

このとき，積分の部分にのみ注目し，一般二項定理を使う．

一般二項定理は実数$\alpha$に対して次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        (1 + x)^{\alpha} = 1 + \alpha x + \cfrac{\alpha(\alpha - 1)}{2!} x^2 + \cfrac{\alpha(\alpha - 1)(\alpha -2)}{3!} x^3 + ・・・
    \end{align}
$$
</span>

すると，流量を求めることができる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\int_0^{\pi} sin^2 \theta \left( 1 - \cfrac{r}{h} cos \theta \right)^{1 / 2} d \theta \\
        &= \int_0^{\pi} sin^2 \theta \left\{1 - \cfrac{1}{2} \cfrac{r}{h} cos \theta - \cfrac{1}{8} \left(\cfrac{r}{h} cos \theta \right)^2 - \cfrac{1}{16} \left(\cfrac{r}{h} cos \theta \right)^3 - ・・・ \right\}d \theta \\
    \end{align}
$$
</span>

ここで，倍角の公式

<span style="font-size:150%;">
$$
    \begin{align}
        &\int sin^2 \theta d \theta = \int \cfrac{1 - cos2 \theta}{2} d \theta \\
        &= \cfrac{1}{2} \theta - \cfrac{1}{4} sin 2 d \theta + C
    \end{align}
$$
</span>

倍角の公式を利用し計算すると，

<span style="font-size:150%;">
$$
    \begin{align}
        &= \cfrac{1}{2} \pi - \cfrac{\pi}{64} \left(\cfrac{r}{h} \right)^2 - \cfrac{5 \pi}{2048} \left(\cfrac{r}{h} \right)^4 - ・・・ \\
        &Q = C \left\{1 - \cfrac{1}{32} \left(\cfrac{r}{h} \right)^2 - \cfrac{5}{1024} \left(\cfrac{r}{h} \right)^4 - ・・・ \right\} \pi r^2 \sqrt{2gh} \\
        &Q = C' \pi r^2 \sqrt{2gh}
    \end{align}
$$
</span>

#### ④三角形大型オリフィス

三角形大型オリフィスは，まず幅$b(z)$を比を使ってもとめる．

接近流速を考慮しない．

<span style="font-size:150%;">
$$
    \begin{align}
        &h_0 : b_0 = (h + h_0 - z) : b(z) \\
        &b(z) = \cfrac{b_0}{h_0} (h + h_0 - z)
    \end{align}
$$
</span>

次に連続の式を立てる．

すると，流量が求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        Q &= C_v C_c \sqrt{2g} \int_h^{h + h_0} \cfrac{b_0}{h_0} (h + h_0 - z) \sqrt{z} dz \\
          &= C \sqrt{2g} \cfrac{b_0}{h_0} \left[ \cfrac{2}{3} h z^{3 / 2} + \cfrac{2}{3} h_0 z^{3/2} - \cfrac{2}{5}  z^{5/2} \right]_h^{h + h_0} \\
          &= C \sqrt{2g} \cfrac{b_0}{h_0} \left\{ \cfrac{4}{15} (h + h_0)^{5/2} - \cfrac{2}{3}(h + h_0)^{3/2} + \cfrac{2}{5} h^{5/2} \right\}
    \end{align}
$$
</span>

#### ⑤潜りオリフィス

潜りオリフィスは完全潜りオリフィスと不完全潜りオリフィスの２つに分類することができる．

完全潜りオリフィスは出口の全断面が水中にある場合を指し,不完全潜りオリフィスは断面の一部が水中にある場合をいう．

完全オリフィスについて考える．

まずは，ベルヌーイの定理から速度を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + f \cfrac{v_2^2}{2g} \\
        &\cfrac{v_1^2}{2g} + h_1 + 0 = \cfrac{v_2^2}{2g} + h_2 + 0 + \cfrac{v_2^2}{2g} \\
        &v_2^2(1 + f) = 2g \left(h_1 - h_2 + \cfrac{v_1^2}{2g} \right) \\
        &v_2 = \cfrac{1}{\sqrt{1 + f}} \sqrt{2g \left(h + \cfrac{v_1^2}{2g} \right)} \\
        &v_2 = C_v \sqrt{2g(h + h_a)}
    \end{align}
$$

</span>

そのあと，連続の式から流量を求める．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q = Av = a_0 = C_c C_v a \sqrt{2g(h + h_a)} = Ca \sqrt{2g(h + h_a)}
    \end{align}
$$

</span>

また，接近流速を考慮しない場合は次のようになる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q = Ca \sqrt{2gh}
    \end{align}
$$

</span>

不完全潜りオリフィスは流量を二つに分けて式を作る．

流量$Q_1$と$Q_2$はこれまでのオリフィスの式を流用し，それらを合計することで流量が求まる．

また，実験値から流量係数$C_1$は0.62，$C_2$は0.53程度の値になる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q = Q_1 + Q_2 = C_1 \sqrt{2g} \int_{h_1}^h \sqrt{z + h_a} b(z)dz + C_2 \sqrt{2g(h - h_a)}
    \end{align}
$$

</span>

### 1.4.2. オリフィスの排水時間

小型オリフィスにおける水を排水する時間を求める．

排水すると水位は下がるので，流速はだんだん小さくなる．

流速が小さくなるということは，流量も小さくなるということであるので，排水時間は長時間かかることになっている．

4.1.オリフィスで誘導した流量の式に$dt$をかけ，排水時間を求めていく．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &Q = Ca \sqrt{2g(z + h_a)} \\
        &Qdt = Ca \sqrt{2g(z + h_a)} dt
    \end{align}
$$

</span>

このとき，$Qdt$は水の体積を表しているので，次式のように置き換えることができる．

このとき，体積は減少しているので符号をつける必要がある．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &V = Ca \sqrt{2g(z + h_a)} dt \\
        &-Adz = Ca \sqrt{2g(z + h_a)} dt \\
        &dt = - \cfrac{A(z)}{Ca \sqrt{2g(z + h_a)}} dz
    \end{align}
$$

</span>

上式を積分することでオリフィスの排水時間が求まる．

<span style="font-size:150%;">
    
$$
    \begin{align}
         t &= \cfrac{-1}{Ca\sqrt{2g}} \int_{h_1}^{h_2} \cfrac{A(z)}{\sqrt{z + h_a}} dz \\
           &= - \cfrac{A(z)}{Ca\sqrt{2g}} \left[2(z + h_a)^{1/2} \right]_{h_1}^{h_2} \\
           &= - \cfrac{2 A(z)}{Ca\sqrt{2g}} \left\{ (h_2 + h_a)^{1/2} - (h_1 + h_a)^{1/2}\right\} \\
           &= \cfrac{2 A(z)}{Ca\sqrt{2g}} \left\{ (h_1 + h_a)^{1/2} - (h_2 + h_a)^{1/2}\right\}
    \end{align}
$$

</span>

また，接近流速を考慮しないとすると，次のようになる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        t = \cfrac{2 A(z)}{Ca \sqrt{2g}} (h_1^{1/2} - h_2^{1/2})
    \end{align}
$$

</span>

下図のような深さ5[m]のところに直径0.2[m]の水栓がついたタンクがある．

この水栓を開いてから水位が4[m]下がるまでの時間を求めよ．

ただし，流量係数は0.6，断面積は400$[m^2]$とする．

<span style="font-size:150%;">
    
$$
    \begin{align}
        t &= \cfrac{2 A(z)}{Ca \sqrt{2g}} (h_1^{1/2} - h_2^{1/2}) \\
          &= \cfrac{2 \times 400}{0.6 \times \cfrac{\pi \times 0.2^2}{4} \times \sqrt{2 \times 9.8}} (5.0^{1/2} - 1.0^{1/2}) \\
          &= 11856[sec] = 3.29[hour]
    \end{align}
$$

</span>

### 1.4.3. 水門

河川や水路を横断する扉を設け，これを上下することにより流量を調節する設備を**水門**という．

水門から出る流れは，下流水位が低い場合の自由流出，高い場合の潜り流出の２つに分けられる．　

水門から出た流速がほぼ一様になるところでベルヌーイの定理を適応する．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + f \cfrac{v_2^2}{2g} \\
        &\cfrac{v_1^2}{2g} + h_1 + 0 = \cfrac{v_2^2}{2g} + h_2 + 0 + \cfrac{v_2^2}{2g} \\
        &v_2^2(1 + f) = 2g (H - h_2) \\
        &v_2 = \cfrac{1}{\sqrt{1 + f}} \sqrt{2g (H - h_2)} \\
        &v_2 = C_v \sqrt{2g(H - h_2)}
    \end{align}
$$

</span>

求めた流速を連続の式に代入する．

このとき，水門の幅は$b$と置く．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &Q = Av = C_c C_v ab \sqrt{2g(H - h_2)} = Cab \sqrt{2g(H - h_2)} \\
    \end{align}
$$

</span>

一般的に接近流速は無視されるときが多いため，考慮しない場合の流量は次式になる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &Q = Cab \sqrt{2g(h_1 - h_2)} \\
    \end{align}
$$

</span>

この式は，自由流出，潜り流出どちらにも適応できる式となっている．

また，上流側水深を測定することにより流量を知ることができる簡易式もある．

簡易式による流量は，流量係数が様々な影響により，値が変わりやすいことを鑑みて，新しい流量係数を使ってもとめる．

また，新しい流量係数はヘンリーの実験結果から簡単に算出することができる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &Q = C'ab \sqrt{2gh_1} \\
    \end{align}
$$

</span>

### 1.4.4. 堰

私たちが普通目にするダムは貯水ダムと呼ばれ，水をためる為に作られている．

高さ15[m]以上のものを貯水ダムといい，それ以下のものを**堰**という．

この堰を越流する水脈を**ナップ**とよび，ナップは完全ナップ，不完全ナップ，付着ナップの３つに分けられる．

完全ナップは堰を超えた水脈が壁を伝わずに自由落下する状態である．

一方，付着ナップは水脈が壁に付着しながら流れていく状態を指す．

不完全ナップは完全ナップと付着ナップの間の状態のことである．

付着ナップは越流水深が小さい場合や下流側水深が高い場合に起きやすく，完全ナップより流量が増加することが知られており，レーボックの実験によれば付着ナップの流量は完全ナップの流量の1.3倍程度になる．

#### ①四角堰

**四角堰**は下図のような堰を指し，四角堰の流量は接近流速を考慮しない長方形大型オリフィスの流量の式を$h_1 = 0$，$h_2 = h$と置くことで求めることができる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q &= Cb \sqrt{2g} \int_0^h \sqrt{z}dz \\
          &= \cfrac{2}{3} Cb \sqrt{2g} h^{3 / 2}
    \end{align}
$$

</span>

JISの四角堰公式（板谷・手島の式)は次のようになる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &Q = Kbh^{3/2} \\
        &K = 1.785 + \cfrac{0.00295}{h} + 0.237 \cfrac{h}{h_d} - 0.428 \sqrt{\cfrac{(B - b)h}{h_d B}} + 0.034 \sqrt{\cfrac{B}{h_d}}
    \end{align}
$$

</span>

この式が使える適応範囲は，$B=0.5～6.3[m]$，$b=0.15～5.0[m]$，$h_d=0.15～0.3[m]$，$bh_d/B^2 \geqq 0.06$，$0.03 \leqq h \leqq \sqrt{b}$である．

$h_d$は**死水深**とよぶ．

#### ②全幅堰

四角堰のような切欠けがなく，越流幅が水路幅に等しい堰を**全幅堰**という．

JISの全幅堰（石原・井田の式）は次のようになる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &Q = Kbh^{3/2} \\
        &K = 1.785 + \left(\cfrac{0.00295}{h} + 0.237 \cfrac{h}{h_d} \right)(1 + \varepsilon) \\
        &h_d \leq 1m, \varepsilon = 0 \\
        &h_s > 1m, \varepsilon = 0.55(h_d - 1)
    \end{align}
$$

</span>

この式が使える適応範囲は，$h \leqq h_d$かつ$h \leqq B / 4$のとき，

$B \geqq 0.5[m]$，$0.3 \leqq h_d \leqq 2.5[m]$，$0.03 \leqq h \leqq 0.8[m]$である．

#### ③三角堰

**三角堰**の流量は三角形大型オリフィスの式を$h_1 = 0$，$h_2 = h$と置くことで求めることができるが，先に水面幅をもとめておく．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &tan \cfrac{\theta}{2} = \cfrac{b(z) / 2}{h - z} \\
        &b(z) = 2(h - z) tan \cfrac{\theta}{2}
    \end{align}
$$

</span>

この水面幅の式を流量の式に代入する．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q &= 2C tan \cfrac{\theta}{2} \sqrt{2g} \int_0^h (h - z) \sqrt{z} dz\\
          &= 2C tan \cfrac{\theta}{2} \sqrt{2g} \left[ \cfrac{2}{3} hz^{3/2} - \cfrac{2}{5} z^{5/2} \right]_0^h \\
          &= 2C tan \cfrac{\theta}{2} \sqrt{2g} \left( \cfrac{2}{3} h^{5/2} - \cfrac{2}{5}h^{5/2} \right) \\
          &= \cfrac{8}{15} C tan \cfrac{\theta}{2} \sqrt{2g} h^{5/2}
    \end{align}
$$

</span>

また，直角三角堰の場合，$\theta = 90°$のため$tan(\theta / 2) = 1$となる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q &= \cfrac{8}{15} C \sqrt{2g} h^{5/2}
    \end{align}
$$

</span>

JISの直角三角堰（沼知・黒川・淵沢の式）は次のようになる．

<span style="font-size:150%;">
    
$$
    \begin{align}
         &Q = K h^{5/2} \\
         &K = 1.354 + \cfrac{0.004}{h} + \left(0.14 + \cfrac{0.2}{h_d} \right) \left(\cfrac{h}{B} - 0.09 \right)^2
    \end{align}
$$

</span>

この式が使える適応範囲は，$h \leqq B / 3$のとき
$0.5 \leqq B \leqq 1.2[m]$，$0.1 \leqq h_d \leqq 0.75 [m]$，$0.07 \leqq h \leqq 0.26[m]$である．


#### ④台形堰

下図のような**台形堰**を考える．

すると台形堰は四角堰と三角堰の組み合わせで流量を求めることができる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q &= \cfrac{2}{3} C_1 b_1 \sqrt{2g} h^{3 / 2} + \cfrac{8}{15} C_2 tan \theta \sqrt{2g} h^{5/2}
    \end{align}
$$

</span>

<span style="font-size:150%;">
    
$
    \begin{align}
        tan \theta = \cfrac{b_2}{h}より
    \end{align}
$

</span>

<span style="font-size:150%;">
    
$$
    \begin{align}
        Q &= \cfrac{2}{3} C_1 b_1 \sqrt{2g} h^{3 / 2} + \cfrac{8}{15} C_2 b_2 \sqrt{2g} h^{3/2} \\
          &= \cfrac{2}{15} C (5 b_1 + 4 b_2) \sqrt{2g} h^{3/2}
    \end{align}
$$

</span>

#### ⑤広頂堰

堰の断面が台形状であり，越流水が堰の面に沿って流れ落ちる場合を**広頂堰**という．

別の言い方をすれば，下流側水深が堰の頂面より低い場合を広頂堰という．

2点間でベルヌーイの定理を適応させる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + f \cfrac{v_2^2}{2g} \\
        &\cfrac{v_1^2}{2g} + h_1 = \cfrac{v_2^2}{2g} + h_2 + \cfrac{v_2^2}{2g} \\
    \end{align}
$$

</span>

このとき，左辺は全水頭と同じになるので，式変形を行い速度及び流量を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + f \cfrac{v_2^2}{2g} = H \\
        &\cfrac{v_2^2}{2g} (1 + f) = H - h_2 \\
        &v_2 = \cfrac{1}{\sqrt{1 + f}} \sqrt{2g(H - h_2)} \\
        &v_2 = C_v \sqrt{2g(H - h_2)} \\
        &Q = Av = C_c C_v b h_2 \sqrt{2g(H - h_2)} = Cbh_2 \sqrt{2g(H - h_2)}
    \end{align}
$$

</span>

さらに，流量を最大にさせる水深$h_2$について考える．

流量を最大にする水深$h_2$は**ベランジェの定理**より次のように定義される．

また，流量が最大となるときの式が求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        &h_2 = \cfrac{2}{3} H \\
        &Q = \cfrac{2}{3} C b H \sqrt{\cfrac{2}{3} gH}
    \end{align}
$$

</span>

#### ⑥潜り堰

下流側水深が堰の頂面より高い場合を**潜り堰**という．

下流の水位が比較的低い場合を完全越流潜り堰といい，堰頂部をこえると射流になるため下流の水位の影響を受けなくなる．

そのため，広頂堰として扱うことができる．

一方，下流の水位が比較的高い場合を潜り堰といい，堰の真上でも常流となるために下流の水位の影響をもろに受ける．

潜り堰における3点間にベルヌーイの定理を適応する．

そのあと連続の式で流量を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{v_1^2}{2g} + h_1 = \cfrac{v_2^2}{2g} + h_2 + \cfrac{v_2^2}{2g} = \cfrac{v_3^2}{2g} + h_3 + \cfrac{v_3^2}{2g} = H\\
        &v_2 = C_v \sqrt{2g(H - h_2)} \\
        &Q = Av = C_c C_v b h_2 \sqrt{2g(H - h_2)} = Cbh_2 \sqrt{2g(H - h_2)} \\
        &Q = Cbh_3 \sqrt{2g(H - h_3)}
    \end{align}
$$

</span>

高潮浸水計算時の越波・越流状態に関して本間の流量公式がある．

完全越流

<span style="font-size:150%;">
$$
    \begin{align}
        &Q = 0.35 bh_1 \sqrt{2gh_1}
    \end{align}
$$

</span>

潜り越流

<span style="font-size:150%;">
$$
    \begin{align}
        &Q = 0.91 bh_2 \sqrt{2g(h_1 - h_2)}
    \end{align}
$$

</span>

下図のような台形堰がある．$b_1 = 0.3[m]$，$b_2 = 0.05[m]$，$h = 0.2[m]$のときの流量を求める．

ただし，流量係数は0.6とする．

<span style="font-size:150%;">
$$
    \begin{align}
        Q &= \cfrac{2}{15} C (5 b_1 + 4 b_2) \sqrt{2g} h^{3/2} \\
          &= \cfrac{2}{15} \times 0.6(5 \times 0.3 + 4 \times 0.05) \sqrt{2 \times 9.8} \times 0.2^{3/2} \\
          &= 0.0539[m/s^3]
    \end{align}
$$

</span>

## 1.5. 管水路の流れ

### 1.5.1. 層流の流速分布

管水路の層流の流れを観察すると，下図のように管の中心で流速が最も大きく，管壁に近づくほど流速は小さくなり，壁上で流速は０となる．

これは，水が粘性を持っているために摩擦応力が働くためである．

このとき，**せん断応力**の大きさから摩擦速度を求める．

せん断応力とは，物体内部のある面の平行方向にすべらせるように作用する応力のことであり，仮想断面を想定してそこに作用する摩擦応力を指す．

まずは，管水路内に円柱型の仮想断面を想定し，釣り合い方程式を立てる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\pi r^2(p_1 - p_2) = \tau \times 2 \pi rl \\
        &\tau = \cfrac{\pi r^2(p_1 - p_2)}{2 \pi rl} = \cfrac{r(p_1 - p_2)}{2l}
    \end{align}
$$

</span>

このとき，$\tau$はせん断応力$[N/m^2]$である．

釣り合いの式から仮想断面におけるせん断応力を求めることができる．

また，管水路の壁面で起きている摩擦応力を式で表わすと次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\tau_0 = \cfrac{r_0(p_1 - p_2)}{2l}
    \end{align}
$$

</span>

このとき，$\tau_0$は摩擦応力$[N/m^2]$である．

つぎにせん断応力を摩擦応力で表わす．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{\tau}{\tau_0} = \cfrac{r(p_1 - p_2)/2l}{r_0(p_1 - p_2)/2l} = \cfrac{r}{r_0} \\
        &\tau = \cfrac{r}{r_0} \tau_0
    \end{align}
$$

</span>

この式により，せん断応力は管中心からの距離と摩擦応力を測定することで計算できる．

また，せん断応力は管中心で最小値0となり，管壁面で最大値$\tau_0$となる直線分布であることがわかる．

さらに，この関係は層流，乱流のいずれの場合においても成立する．

水力学の分野では，管壁面の摩擦応力から**摩擦速度**という物理量が導かれる．

摩擦速度は間接的に摩擦応力の大きさを表しており，速度の無次元化を行う際に利用される．

無次元のパラメータとしてはレイノルズ数やフルード数があげられる．

これら係数のように無次元化することにより，流れ状況がより把握しやすくなる．

摩擦速度もその無次元化するために必要なパラメータのひとつである．

摩擦速度はせん断応力を密度で割り，1 / 2乗することで求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        &u_* = \sqrt{\cfrac{\tau_0}{\rho}}
    \end{align}
$$

</span>

このとき，$u_*$は摩擦速度$[m/s]$である．

このとき，速度の記号が$v$から$u$に変化している．

私たちが一般に言っている速度は**ラグランジュ速度**といい，速度の値は時間変数だけで決まる．

一方，時間変数と位置変数によって変化する速度を**オイラー速度**という．

とくに断りがない限り，ラグランジュ速度は$v(t)$，オイラー速度は$u(x,y,z,t)$の記号を用いて表現する．

今までは，平均流速を考えていたために記号$v$を使用していましたが，今考えている摩擦速度は粘性があるために位置によって速度が変化する．

そのため，記号$u$を使用する．

記号$v$は平均流速，記号$u$は測定値で考えるといい．

では，管水路における層流の流速分布と乱流の流速分布を求める．

層流の場合，壁面近くのせん断応力は管水路の半径が増加すると流速は減少するために負の符号をつける必要がある．

<span style="font-size:150%;">
$$
    \begin{align}
        &\tau = - \mu \cfrac{du}{dr}
    \end{align}
$$

</span>

層流における管水路の流速はせん断応力の式から釣り合いの式を立て，積分することにより求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &- \mu \cfrac{du}{dr} = \cfrac{r(p_1 - p_2)}{2l} \\
        & du = - \cfrac{r(p_1 - p_2)}{2 \mu l} dr \\
        & u = - \int \cfrac{r(p_1 - p_2)}{2l \mu} dr \\
        &    = - \cfrac{r^2(p_1 - p_2)}{4 \mu l} + C
    \end{align}
$$

</span>

このとき，境界条件より$r = r_0$のとき，$u = 0$となるので，積分定数が求まる．

また，積分定数を速度の式に代入することで層流の流速分布が求まる．

層流の流速分布は二次の放物線で表わされる．

<span style="font-size:150%;">
$$
    \begin{align}
        & C = \cfrac{r_0^2(p_1 - p_2)}{4 \mu l} \\
        & u = - \cfrac{r^2(p_1 - p_2)}{4 \mu l} + \cfrac{r_0^2(p_1 - p_2)}{4 \mu l} \\
        &    = \cfrac{(r_0^2 - r^2)(p_1 - p_2)}{4 \mu l}
    \end{align}
$$

</span>

最後に層流の全流量を求める．

まず，下図で色のついている微小面積について誘導する．

<span style="font-size:150%;">
$$
    \begin{align}
        &dA = \pi (r + dr)^2 - \pi r^2 = 2 \pi r dr + \pi dr^2 = 2 \pi r dr
    \end{align}
$$

</span>

$\pi dr^2$は微小距離を２乗しているのでさらに小さい値になり，0とみなすことができる．

次に，連続の式から流量を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        q &= udA \\
          &= \cfrac{(r_0^2 - r^2)(p_1 - p_2)}{4 \mu l} \times 2 \pi r dr \\
          Q &= \int_0^{r_0} \pi r \cfrac{(r_0^2 - r^2)(p_1 - p_2)}{2 \mu l} dr \\
          &= \cfrac{\pi(p_1 - p_2)}{2 \mu l} \left[ \cfrac{r_0^2}{2}r^2 - \cfrac{1}{4}r^4 \right]_0^{r_0} \\
          &= \cfrac{\pi(p_1 - p_2)r_0^4}{8 \mu l}
    \end{align}
$$

</span>

この流量を表す式を**ハーゲン・ポアズイユの式**といい，層流における流量を示す．

ハーゲン・ポアズイユの式は粘性係数の測定などに使われており，様々な条件の下で成り立ってる．

①乱れ変動がなくレイノルズ応力が０である．（つまり層流である．）

②定常流れである．（時間的に変化しない）

③断面方向に流れない

④流体は連続体としてふるまう

⑤壁面において流体の速度0である．

さらに，平均流速は次のように表される．

<span style="font-size:150%;">
$$
    \begin{align}
        v &= \cfrac{Q}{\pi r_0^2} \\
          &= \cfrac{\pi(p_1 - p_2)r_0^4}{8 \mu l} \times \cfrac{1}{\pi r_0^2} \\
          &= \cfrac{(p_1 - p_2)r_0^2}{8 \mu l}
    \end{align}
$$

</span>

### 1.5.2. 乱流の流速分布

乱流は層流とは違い水粒子が入り混じって流れる．

そのため，壁面にごく近いところを除いて，下図のように流速が一様に近い分布となっている．

乱流中は速度に対し不規則な変動を伴っており，流れを平均値$\bar{u}, \bar{v}$と変動値$u', v'$に分けて方程式を作る必要がある．（レイノルズ分解）

流れ方向の瞬間速度$u$と直角方向の瞬間速度$v$は以下のようにあらわされる．

<span style="font-size:150%;">
$$
    \begin{align}
        &u = \bar{u} + u' \\
        &v = \bar{v} + v'
    \end{align}
$$

</span>

乱流時のせん断応力は流体の粘性による応力（摩擦応力）に**レイノルズ応力**が加わる．

レイノルズ応力は数学者**ブーシネスク**に乱流によって生じるせん断応力と定義された．

レイノルズ応力を誘導する．

まずは流れ方向のせん断応力を運動量方程式から考える．

このとき単位時間をかけておく必要がある．

<span style="font-size:150%;">
$$
    \begin{align}
        &F = ma = \rho V a = \rho V v'/dt = \rho Q v' \\
        &\tau dt = \cfrac{F}{A}dt = \cfrac{\rho Q v'}{A}dt = \rho u v' dt = \rho v'(\bar{u} + u')dt
    \end{align}
$$

</span>

このとき，それぞれの平均流速は次のように表すことができる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\bar{u} = \cfrac{1}{t} \int_0^t u dt \\
        &\bar{u'} = \cfrac{1}{t} \int_0^t u' dt = 0\\
        &\bar{v'} = \cfrac{1}{t} \int_0^t v' dt = 0
    \end{align}
$$

</span>

変動速度の平均が０になる理由

<span style="font-size:150%;">
$$
    \begin{align}
        \bar{u'} &= \cfrac{1}{t} \int_0^t u' dt \\
                 &= \cfrac{1}{t} \int_0^t (u - \bar{u}) dt \\
                 &= \cfrac{1}{t} \int_0^t u dt -  \cfrac{1}{t} \int_0^t \bar{u} dt \\
                 &= \bar{u} - \bar{u} = 0
    \end{align}
$$

</span>

レイノルズ応力を表す式はせん断応力の式を定積分することで求められる．

<span style="font-size:150%;">
$$
    \begin{align}
        \tau \int_0^t 1 dt &= \int_0^t \rho v'(\bar{u} + u')dt \\
        \tau &= \cfrac{1}{t} \rho \int_0^t (\bar{u} v' + u' v') dt \\
             &= \rho \bar{u} \cfrac{1}{t} \int_0^t v'dt + \rho \cfrac{1}{t} \int_0^t u'v' dt \\
             &= 0 + \rho \overline{u' v'} \\
             &= \rho \overline{u' v'}
    \end{align}
$$

</span>

また，抵抗は流れの方向と反対に働くので，負の符号をつける必要がある．

この式がレイノルズ応力を表す式となる．

<span style="font-size:150%;">
$$
    \begin{align}
        \tau = - \rho \overline{u' v'}
    \end{align}
$$

</span>

乱流中の粘性を考慮した全せん断応力は次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        \tau =  \left(\mu \cfrac{du}{dz} - \rho \overline{u' v'} \right)
    \end{align}
$$

</span>

十分に発達した乱流中では，壁面近くを除くとレイノルズ応力が摩擦応力よりはるかに大きくなることがわかる．

そこで，摩擦応力は無視できるとし，レイノルズ応力についてのみ考える．

乱流中の流速の変動値の要因は流速差（速度勾配）である．

したがって乱れの大きさは平均流速の空間的な勾配に比例すると考えられる．

よって$u'$は以下のように記述できる．

<span style="font-size:150%;">
$$
    \begin{align}
        u' = l' \cfrac{d \bar{u}}{dz}
    \end{align}
$$

</span>

このとき，$l'$は速度勾配を速度にするための長さスケールもしくは渦の大きさ（直径）[m]である．

また，乱流中の流速の変動値$v'$が$u'$に比例していると仮定し，その比例定数を$a$とおく．

すると，次のような式ができる．

<span style="font-size:150%;">
$$
    \begin{align}
        u'v' = al'^2 \left( \cfrac{d \bar{u}}{dz} \right)^2
    \end{align}
$$

</span>

ここで流体塊の混合運動のスケールに相当する混合距離$l$を導入する．これをプラントルの混合距離の仮説という．

<span style="font-size:150%;">
$$
    \begin{align}
        l^2 &= al'^2 \\
        \eta &= \rho l^2 \cfrac{d \bar{u}}{dz} \\
        \epsilon &= l^2 \cfrac{d \bar{u}}{dz} = \cfrac{\eta}{\rho} \\
        \tau &= - \rho \overline{u' v'} = \rho l^2 \left(\cfrac{d \bar{u}}{dz} \right)^2 = \eta \cfrac{d \bar{u}}{dz}
    \end{align}
$$

</span>

このとき，$l$は混合距離[m]，$\eta$は渦粘性係数$[Pa \cdot s]$，$\epsilon$は渦動粘性係数$[m^2/s]$である．

渦粘性係数や渦動粘性係数は流体の性質に関係なく，流速や場所によって異なる係数となる．

プラントルとカルマンは壁面近傍の混合距離は高さに比例すると仮定した．

壁面では速度が０となるため，渦が発生しない．

<span style="font-size:150%;">
$$
    \begin{align}
        l = \kappa z
    \end{align}
$$

</span>

このとき，$\kappa$はカルマン定数[単位なし]である．

カルマン定数は実験により値を求めることができ，通常0.4くらいとされている．

また，壁面近傍のせん断応力は$\tau = \tau_0$とおくことができる．

<span style="font-size:150%;">
$$
    \begin{align}
        \tau_0 = \rho l^2 \left(\cfrac{d \bar{u}}{dz} \right)^2 = \rho \kappa^2 z^2 \left(\cfrac{d \bar{u}}{dz} \right)^2
    \end{align}
$$

</span>

摩擦速度を上式に適応させ，積分することで管内の流速分布を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        u_* &= \sqrt{\cfrac{\tau_0}{\rho}} \\
        u_*^2 \rho &= \rho \kappa^2 z^2 \left(\cfrac{d \bar{u}}{dz} \right)^2 \\
        \cfrac{du}{dz} &= \cfrac{u_*}{\kappa} \cfrac{1}{z} \\
        u &= \cfrac{u_*}{\kappa} \int \cfrac{1}{z} dz = \cfrac{u_*}{\kappa} ln z + C \\
          &= 2.30 \cfrac{u_*}{\kappa} log z + C = 5.75 u_* log z + C
    \end{align}
$$

</span>

管の中心$z = r_0$のとき最大流速$u = u_0$とおいて積分定数を求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &u_0 = 5.75 u_* log r_0 + C \\
        &C = u_0 - 5.75 u_* log r_0
    \end{align}
$$

</span>

積分定数を管内の流速の式に代入する．

<span style="font-size:150%;">
$$
    \begin{align}
        &u = 5.75 u_* log z + u_0 + 5.75 u_* log r_0\\
        &\cfrac{u - u_0}{u_*} = 5.75 log z - 5.75 log r_0 \\
        &\cfrac{u_0 - u}{u_*} = 5.75 log r_0 - 5.75 log z = 5.75 log \cfrac{r_0}{z}
    \end{align}
$$

</span>

この式は**速度欠損則**と呼ばれており，乱流の流速分布が対数で表わせることを示す．

しかし，最大流速がわからないと流速を求めることができないので実用的な式とは言えない．

そこで，プラントルとカルマンは，乱流の流速分布式として以下の式を誘導する．

便宜的な境界条件として($z = z_0$において$u = 0$)を導入する．

<span style="font-size:150%;">
$$
    \begin{align}
        &0 = 5.75 u_* log z_0 + C \\
        &C = - 5.75 u_* log z_0
    \end{align}
$$

</span>

よって，

<span style="font-size:150%;">
$$
    \begin{align}
        &u = 5.75u_* logz - 5.75 u_*log z_0 \\
        &\cfrac{u}{u_*} = 5.75 log \cfrac{z}{z_0}
    \end{align}
$$

</span>

流速の絶対値を決めるうえで，具体的に$y_0$を決定する必要がある．

$y_0$は壁面近傍の物理量で決定される．

壁面が滑らかな場合は，壁面せん断力$\tau_0$，流体の密度$\rho$，流体の動粘性係数$\nu$で決定される．

次元解析（物理量における，長さ，質量などの次元から複数の物理量の間の関係を予測）より

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{z_0 \sqrt{\tau_0 / \rho}}{\nu} = \cfrac{z_0 u_*}{\nu} = const.\\
    \end{align}
$$

</span>

この式の関係を壁法則という．

これを代入すれば次式を得る．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{u}{u_*} = 5.75 log \cfrac{u_* z}{\nu} + A
    \end{align}
$$

</span>

この式は，乱流における流速分布を表す一般式であり，**速度分布の対数法則**と呼ばれる．

また，ニクラーゼの実験結果から$A$の値を求めることができる．

<span style="font-size:150%;">
$$
    \begin{align}
        &滑面領域 : \cfrac{u_*k}{\nu} \leq 5, A = 5.50 + 5.75 log \cfrac{u_* k}{\nu} \\
        &遷移領域 : 5 < \cfrac{u_* k}{\nu} \leq 70 , A = f \left(\cfrac{u_*k}{\nu} \right)\\
        &粗面領域 : 70 < \cfrac{u_* k}{\nu} , A = 8.50\\
    \end{align}
$$

</span>

このとき，$k$は壁面の粗さ（凹凸）の平均高さ（絶対粗度）[m]である．

壁面が粗い場合は，次元解析より

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{z_0}{k} = f\left(\cfrac{u_* k}{\nu} \right) = const.\\
    \end{align}
$$

</span>

これを代入すれば次式を得る．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{u}{u_*} = 5.75 log \cfrac{z}{k} + A
    \end{align}
$$

</span>

これらの式をそれぞれの流速分布の式に代入する．

<span style="font-size:150%;">
$$
    \begin{align}
        &滑面領域 : \cfrac{u}{u_*} = 5.50 + 5.75 log \cfrac{u_* z}{\nu} \\
        &遷移領域 : \cfrac{u}{u_*} = f \left(\cfrac{u_*k}{\nu} \right) + 5.75 log \cfrac{z}{k}\\
        &粗面領域 : \cfrac{u}{u_*} = 8.50 + 5.75 log \cfrac{z}{k}\\
    \end{align}
$$

</span>

これらの式は壁面近傍，たとえば$z = 0$のとき，$u = - \infty$となり，不適切であることがわかる．

また，実際の流れでは下図のように**粘性底層**が存在する．

粘性底層内では，レイノルズ応力項は無視できることから，次のように表現できる．

粘性底層とは壁面近傍における粘性の影響を強く受けた領域のことである．

<span style="font-size:150%;">
$$
    \begin{align}
        &せん断応力分布 : \tau \fallingdotseq \tau_0 = \mu \cfrac{du}{dz} - \rho \overline{u' v'} \\
        &レイノルズ応力 : - \rho \overline{u' v'} = 0
    \end{align}
$$

</span>

したがって

<span style="font-size:150%;">
$$
    \begin{align}
        &せん断応力分布 : \tau \fallingdotseq \tau_0 = \mu \cfrac{du}{dz} 
    \end{align}
$$

</span>

の関係が成立する．

ここで，摩擦速度の式$u_* = \sqrt{\tau_0 /\rho}$を上式に代入すると次の式になる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\rho u_*^2 = \mu \cfrac{du}{dz} \\
        & \cfrac{1}{u_*}du = \cfrac{\rho}{\mu} u_* dz
    \end{align}
$$

</span>

$\nu = \cfrac{\mu}{\rho}$より積分すると次式になる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{u}{u_*} = \cfrac{u_* z}{\nu} + C \\
    \end{align}
$$

</span>

壁面では流速が0（$y = 0$で$u = 0$）となることより積分定数$C$は$C = 0$となる．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{u}{u_*} = \cfrac{u_* z}{\nu} \\
    \end{align}
$$

</span>

滑面領域での流速分布式と粘性底層の式の交点から粘性底層の厚さを求める．

<span style="font-size:150%;">
$$
    \begin{align}
        &\cfrac{u_* \delta}{\nu} = 11.6 \\
        &\delta = \cfrac{11.6u_*}{\nu}
    \end{align}
$$

</span>

このとき，$\delta$は流速分布式と粘性底層の式の交点における粘性底層の厚さ[m]である．

断面全体の流速分布を考慮して乱流の平均流速を求める．

下記の$v$は管全体の平均流速を表す．

まずは連続の式から方程式を立てる．

円の微小面積から考える．

<span style="font-size:150%;">
$$
    \begin{align}
        Q &= \int_0^{r_0} udA \\
          &= \int_0^{r_0} 2 \pi r u_* \left\{5.50 + 5.75 log \cfrac{u_*(r_0 - r) }{\nu} \right\}dr
    \end{align}
$$

</span>

ここで置換積分を行い，logの中の数式を$t$に置き換える．

まずは条件を整えていく．

<span style="font-size:150%;">
$$
    \begin{align}
        &t = \cfrac{u_*(r_0 - r)}{\nu} \\
        &r = 0のとき \\
        &t = \cfrac{u_* r_0}{\nu} \\
        &r = r_0のとき \\
        &t = 0 \\
        &r = - \cfrac{\nu}{u_*}t + r_0 \\
        &dr = - \cfrac{\nu}{u_*}dt
    \end{align}
$$

</span>

これで条件がそろったので，上式を連続の式に代入する．

<span style="font-size:150%;">
$$
    \begin{align}
        Q = 2 \pi u_* \int_0^{r_0} 5.50rdr - 2 \pi \nu \int_{u_* r_0 / \nu}^0 5.75 \left(- \cfrac{\nu}{u_*}t + r_0 \right)logt dt
    \end{align}
$$

</span>

式が難しくなってきたので，第一項と第二項を分けて計算する．

まずは第一項の積分をする．

<span style="font-size:150%;">
$$
    \begin{align}
         2 \pi u_* \int_0^{r_0} 5.50rdr &= 2 \pi u_* \left[ 5.50 \cdot \cfrac{1}{2}r^2 \right]_0^{r_0} \\
        &= 5.50 \pi r_0^2 u_*
    \end{align}
$$

</span>

続いて第二項の積分である．

第二項は部分積分を行う必要がある．

<span style="font-size:120%;">
$$
    \begin{align}
         &- 2 \pi \nu \int_{u_* r_0 / \nu}^0 5.75 \left(- \cfrac{\nu}{u_*}t + r_0 \right)logt dt \\
         &= 2 \pi \nu \int_0^{u_* r_0 / \nu} 5.75 \left(- \cfrac{\nu}{u_*}t + r_0 \right)logt dt \\
         &部分積分より \\
         &= 2 \pi \nu \left[5.75 \left(- \cfrac{\nu}{2 u_*} t^2 + r_0t \right) log_{10} t \right]_0^{u_*r_0/ \nu} - 2 \pi \nu \int_0^{u_*r_0/ \nu} \cfrac{5.75}{ln10} \left(- \cfrac{\nu}{2 u_*} t^2 + r_0t \right) \cfrac{1}{t}dt \\
        &部分積分の第一項は \\
        &2 \pi \nu \left[5.75 \left(- \cfrac{\nu}{2 u_*} t^2 + r_0t \right) log_{10} t \right]_0^{u_*r_0/ \nu} \\
        &= 5.75 \pi r_0^2  u_*log_{10} \cfrac{u_* r_0}{\nu}  \\
        &第二項は \\
        &  - 2 \pi \nu \int_0^{u_*r_0/ \nu} \cfrac{5.75}{ln10} \left(- \cfrac{\nu}{2 u_*} t^2 + r_0t \right) \cfrac{1}{t}dt \\
        &= - 2 \pi \nu \cfrac{5.75}{ln10} \int_0^{u_*r_0/ \nu}  \left(- \cfrac{\nu}{2 u_*} t + r_0 \right) dt \\
        &= - 2 \pi \nu \cfrac{5.75}{ln10} \left[- \cfrac{\nu}{4 u_*} t^2 + r_0t \right]_0^{u_*r_0/ \nu} \\
        &= -  \cfrac{5.75 \times 3}{2ln10} \pi r_0^2 u_* \\
        &= - 3.75 \pi r_0^2 u_*
    \end{align}
$$

</span>

第一項の積分結果と第二項の積分結果を合わせる．

<span style="font-size:120%;">
$$
    \begin{align}
        Q &= 5.50 \pi r_0^2 u_* + 5.75 \pi r_0^2  u_*log_{10} \cfrac{u_* r_0}{\nu} - 3.75 \pi r_0^2 u_* \\
          &= \pi r_0^2 u_* \left(1.75 + 5.75log \cfrac{u_* r_0}{\nu} \right)
    \end{align}
$$

</span>

最後に，流量を断面積で割ると平均流速の式が求まる．

<span style="font-size:150%;">
$$
    \begin{align}
        \cfrac{1}{u_*} \cfrac{Q}{A} &= \cfrac{1}{u_*} \cfrac{\pi r_0^2 u_* \left(1.75 + 5.75log \cfrac{u_* r_0}{\nu} \right)}{\pi r_0^2} \\
        \cfrac{v}{u_*}&= 1.75 + 5.75log \cfrac{u_* r_0}{\nu}
    \end{align}
$$

</span>

これらの計算方法を遷移領域と粗面領域も同様に行う．

すると乱流における平均流速の式が表される．

<span style="font-size:150%;">
$$
    \begin{align}
        滑面領域 : \cfrac{v}{u_*}&= 1.75 + 5.75log \cfrac{u_* r_0}{\nu} \\
        遷移領域 : \cfrac{v}{u_*}&= f \left(\cfrac{u_* k}{\nu} \right) - 3.75 + 5.75 log \cfrac{z}{k} \\
        粗面領域 : \cfrac{v}{u_*}&= 4.75 + 5.75log \cfrac{u_* r_0}{\nu} \\
    \end{align}
$$

</span>

### 1.5.3. 平均流速の公式

平均流速を求める公式はたくさん存在しており，ここではそれら公式について紹介する．

#### ①**ダルシー・ワイズバッハの式**を変形して平均流速を求める方法である．

式にすると次のようになる．

<span style="font-size:150%;">
$$
    \begin{align}
        &h_l = f \cfrac{l}{D} \cfrac{v^2}{2g} \\
        &v = \sqrt{\cfrac{2gDh_l}{fl}} = \sqrt{\cfrac{2}{f}} \sqrt{gDI} \\
        &I = \cfrac{h_l}{l}
    \end{align}
$$

</span>

このとき，$f$は摩擦損失係数[単位なし]，$I$は動水勾配[単位なし]，$h_l$は摩擦損失水頭[m]，$l$は管長[m]，$D$は管径[m]である．

**動水勾配**は単位長さあたりの摩擦損失水頭で表わされる．

また，円管の径深$R$を使って次のようにも表現できる．

<span style="font-size:150%;">
$$
    \begin{align}
        &R = \cfrac{A}{S} = \cfrac{\pi D^2 /4}{\pi D} = \cfrac{D}{4} \\
        &v = \sqrt{\cfrac{2}{f}} \sqrt{gDI} = \sqrt{\cfrac{8}{f}} \sqrt{gRI}
    \end{align}
$$

</span>

上式は合理的で精度も良いが，摩擦損失係数の算定が複雑であり，実際の管水路の計算には若干不便である．

そのため，直接的に平均流速が求められる実用的な式が多数提唱される．

#### ②シェジーの公式

シェジーの公式は1769年にフランスの**アントワーヌ・シェジー**によって提案された実験式である．

<span style="font-size:150%;">
$$
    \begin{align}
        &v = C \sqrt{RI}
    \end{align}
$$

</span>

このとき，$C$はシェジー係数$[m^{1/2}/s]$，$R$は径深，$I$は動水勾配である．

**シェジー係数**は水路の水の流れやすさを表している．

提唱された当時はシェジー係数を定数として扱っていたが，その後実測により定数でないことが証明された．

そのため，現在はあまり使われていない．

#### ③ガンギレー・クッターの公式

ガンギレー・クッターの公式はシェジー係数が定数でないことを証明した．

<span style="font-size:150%;">
$$
    \begin{align}
        &v = \cfrac{23 + 1 / n + 0.00155 / I}{1 + (23 + 0.00155 / I)(n / \sqrt{R})} \sqrt{RI}
    \end{align}
$$

</span>

このとき，$n$は粗度係数$[m^{- 1/3} \cdot s]$である．

**粗度係数**は壁面粗さの程度を与える係数であり，値が大きいほど水路壁面が粗いことを表す．

しかし，この式も複雑すぎたため，現在はほとんど使われていない．

下水道の配水管設計のときに用いられる程度である．

#### ④ヘーゼン・ウィリアムスの公式

ヘーゼン・ウィリアムスの公式は管水路の水の流れを動水勾配によって関係づけた経験式であり，上水道の配水管設計のとき用いられていた．

この式は非常に計算しやすいというメリットがあるが，水にしか適応することができず，水温や粘性について考慮していないことから次第に使われなくなっていった．

ちなみに，式中の$C_H$は管種によって決定する粗度係数を示している．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &v = 0.35464 C_H D^{0.63} I^{0.54}
    \end{align}
$$

</span>

#### ⑤マニングの公式

**マニングの公式**はアイルランドの技術者であった**ロバート・マニング**によって考案された公式であり，現在最もよく使われている平均流速の公式である．

よく用いられている理由としては，式が簡単なことや流速がガンギレー・クッターの公式とほぼ同一であることが挙げられる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &v = \cfrac{1}{n} R^{2/3} I^{1/2}
    \end{align}
$$

</span>

また，マニングの公式とダルシー・ワイズバッハの式から摩擦損失係数$f$と粗度係数$n$との関係式が求まる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &f = \cfrac{8gn^2}{R^{2/3}} = \cfrac{12.7gn^2}{D^{1/3}} = \cfrac{124.5n^2}{D^{1/3}}
    \end{align}
$$

</span>

内径0.6[m]の管水路に動水勾配$1 / 200$で水を流すときの平均流速を求めよ．

ただし，粗度係数は0.013とする．

径深を求めてから，平均流速を計算する．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &R = \cfrac{D}{4} = \cfrac{0.6}{4} =  0.15[m] \\
        &v = \cfrac{1}{n} R^{2 / 3} I^{1 / 2} = \cfrac{1}{0.013} 0.15^{2/3} \cfrac{1}{200}^{1/2} = 1.54[m/s]
    \end{align}
$$

</span>



### 1.5.4. 摩擦損失水頭

管水路に水を流すと２点間には必ず圧力差が発生する．

これは，水が菅内に流れているときに摩擦によってエネルギーが消耗されているからである．

そこで，摩擦による損失水頭を考える必要がある．

下図の**動水勾配線**はエネルギー線より速度水頭分だけ低い位置にあり，管径が変化せず流速の等しい区間ではエネルギー線と動水勾配線は平行となる．

また，管水路の途中で水柱を立てると，水柱は動水勾配線まで上がる．

図を見ると断面②に**摩擦損失水頭**が加わっているのが分かる.

また、管水路は流速分布のある流れなのに対し、ベルヌーイの式には平均流速が用いられる．

そのため，速度水頭の前に**エネルギー係数**$\alpha$をつける必要がある．

エネルギー係数の実用的な値としては1.0または1.1程度が妥当といわれている．

では，ベルヌーイの式を立てていく．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &\alpha \cfrac{v_1^2}{2g} + z_1 + \cfrac{p_1}{\rho g} = \alpha \cfrac{v_2^2}{2g} + z_2 + \cfrac{p_2}{\rho g} + h_l
    \end{align}
$$

</span>

このとき，$\alpha$はエネルギー係数[単位なし]，$h_l$は摩擦損失水頭[m]である．

摩擦損失水頭は速度水頭と管長に比例し，管径に反比例することが分かっている．

そのため，摩擦損失水頭は次式で定義される．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_l = f \cfrac{l}{D} \cfrac{v^2}{2g}
    \end{align}
$$

</span>

このとき，$I$は管長[m]，$D$は管径[m]である．

この式は**ダルシー・ワイズバッハの式**と呼ばれており，**ヘンリー・ダルシー**によって開発され，**ユリウス・ワイズバッハ**によって修正された．

ヘンリー・ダルシーは19世紀フランスの技術者であり，地下水の流れに関するダルシーの法則も発見している．

ダルシー・ワイズバッハの式中にある摩擦損失係数は層流，乱流の滑面領域，乱流の遷移領域，乱流の粗面領域に対し次式で表される．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &層流 : f = \cfrac{64}{R_e} \\
        &乱流の滑面領域 : \cfrac{1}{\sqrt{f}} = 2.03log(R_e \sqrt{f}) - 0.8 \\
        &乱流の遷移領域 : \cfrac{1}{\sqrt{f}} = 1.74 - 2.03log \left(\cfrac{2k}{D} + \cfrac{18.7}{R_e \sqrt{f}} \right) \\
        &乱流の粗面領域 : \cfrac{1}{\sqrt{f}} = 2.03log \cfrac{D}{k} + 1.138
    \end{align}
$$

</span>

しかし，これらの式を使って摩擦損失係数をいちいち計算するのはとても面倒である．

そこで，摩擦損失係数の値を与える実用的なものとして，**ムーディ線図**がある．

ムーディ線図は層流及び乱流についての摩擦損失係数の値を$R_e$と$k / D$を用いて表したものである．

各種管の標準的な$k$の値は下表で与えられる．



<span style="font-size:150%;">
    
$$
    \begin{align}
        ムーディ線図
    \end{align}
$$

</span>

|<span style="font-size:120%;">管種</span>|<span style="font-size:120%;">壁面状態</span>|<span style="font-size:120%;">$k[m]$</span>|
|---|---|---|
|<span style="font-size:120%;">塩化ビニル管</span>|<span style="font-size:120%;">工業的に滑らか</span>|<span style="font-size:120%;">$～2.0 \times 10^{-6}$</span>|
|<span style="font-size:120%;">　　</span>|<span style="font-size:120%;">アスファルト舗装</span>|<span style="font-size:120%;">$(1.0～1.5) \times 10^{-4}$</span>|

直径0.1[m]，長さ50[m]の滑面円管に水を流量$0.002[m^3/s]$で流すと，摩擦損失水頭はいくらになるか．

ただし，動粘度は$1.139 \times 10^{-6}$とする．

まずは，円管内の流れが層流なのか乱流なのかを確かめる必要がある．

レイノルズ数を求めるためには平均速度が必要であるので連続式から平均速度を求めていく．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &v = \cfrac{Q}{A} = \cfrac{Q}{\pi D^2 /4} = \cfrac{0.002}{\pi \times 0.1^2 /4} = 0.26[m/s]
    \end{align}
$$

</span>

速度が求まったので，レイノルズ数を計算する．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &R_e = \cfrac{Dv}{\nu} = \cfrac{0.1 \times 0.26}{1.139 \times 10^{-6}} = 22388
    \end{align}
$$

</span>

レイノルズ数の値が40000より大きいので乱流であることが分かる．

また，円管は滑面なので摩擦損失水頭の式は乱流の滑面領域を使う．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &\cfrac{1}{\sqrt{f}} = 2.03log(R_e \sqrt{f}) - 0.8 \\
        &f = \cfrac{1}{\left\{ 2.03log(R_e \sqrt{f}) - 0.8 \right\}^2} = \cfrac{1}{\left\{ 2.03log(22388 \sqrt{f}) - 0.8 \right\}^2}
    \end{align}
$$

</span>

これをニュートン法で解くと$f$の値は0.0245に収束する．

この値をダルシー・ワイズバッハの式に代入する．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_l = f \cfrac{l}{D} \cfrac{v^2}{2g} = 0.0245 \times \cfrac{50}{0.1} \times \cfrac{0.26^2}{2 \times 9.8} = 0.041[m]
    \end{align}
$$

</span>

別海としてムーディ線図から$f = 0.025$を読み取ることが出来る．

### 1.5.5. 形状損失水頭

管水路が直線で管径が変化しない場合，摩擦損失水頭のみを考えればいいが，実際の管水路はそんなに簡単ではない．

途中で管径が変化したり曲がっていたり，流量調整用の弁がついてたりする．

これらの局所的な変形部分によりエネルギーはどんどん消耗されていく．

ここでは，５つの形状損失水頭を述べる．

#### ①流入による損失水頭

水槽や貯水池から管水路に流入する際に起こる損失のことである．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_e = f_e \cfrac{v^2}{2g}
    \end{align}
$$

</span>

このとき，$h_e$は流入による損失水頭[m]，$f_e$は流入損失係数[単位なし]である．

流入損失係数はワイズバッハの実験結果よりある程度の値がわかっている．

#### ②断面変化による損失水頭

菅水路の断面変化には，急拡，急縮，漸拡，漸縮の４つの形がある．

##### [1] 急拡による損失水頭

断面積が急に拡大すると，流れは急には広がらず拡大部に渦が生じエネルギー損失が起こる．

急拡による損失水頭はボルタの式とも呼ばれ，実測値と十分な精度で一致することが知られている．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_{se} = f_{se} \cfrac{v_1^2}{2g} \\
        &f_{se} = \left(1 - \cfrac{A_1}{A_2}\right)^2
    \end{align}
$$

</span>

このとき，$h_{se}$は急拡による損失水頭[m]，$f_{se}$は急拡損失係数[単位なし]，$A$は断面積$[m^2]$である．

##### [2]　急縮による損失水頭

断面積が急に縮まると流れはいったん収縮し始め，そのあと拡大して流れていく．

急縮は圧力水頭が速度水頭に変化するので，エネルギー損失は無視できるほどに小さい．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_{sc} = f_{sc} \cfrac{v_2^2}{2g} 
    \end{align}
$$

</span>

このとき，$h_{sc}$は急縮による損失水頭[m]，$f_{sc}$は急縮損失係数[単位なし]である．

##### [3]漸拡による損失水頭

漸拡は徐々に管径が広がっている様を表している．

漸拡による損失係数は拡がり角$\theta$と管径比$D_2/D_1$によって決まる係数であり，拡がり角が$5°～6°$のときに最小となる．

この角度を超えると流れが管壁から剥離し，周辺の管壁と流れの間に渦を生じ，値が急増する．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_{ge} = f_{ge} \cfrac{(v_1 - v_2)^2}{2g} = f_{ge} f_{se}\cfrac{v_1^2}{2g}
    \end{align}
$$

</span>

このとき，$h_{ge}$は漸拡による損失水頭[m]，$f_{ge}$は漸拡損失係数[単位なし]である．

漸拡損失係数はギブソンの実験値から求まる．

##### [4]漸縮による損失水頭

一般に断面が流れ方向に収縮していく場合は，渦が発生しにくく流れが安定している．

したがって，断面が収縮することに起因するエネルギー損失はきわめて少ない．

そのため，漸縮が緩やかなときは漸縮損失水頭を無視する場合もある．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_{gc} = f_{gc} \cfrac{v_2^2}{2g} 
    \end{align}
$$

</span>

このとき，$h_{gc}$は漸縮損失水頭[m]，$f_{gc}$は漸縮損失係数[単位なし]である．

漸縮損失係数は下図から求められる．

#### ③曲がりによる損失水頭

管水路の変化部は，次第に方向を変えるものを曲がり，

急に折れ曲がるものを屈折と定義されている．

##### [1] 曲がりによる損失水頭

曲がりによる損失水頭は，曲率半径$\rho$と管径$D$，曲がりの中心角$\theta$が関係している．

曲がりによる損失水頭は曲がりの内側で流れが剥離し，向心力の影響で渦が発生する．

曲がりは複雑な様相を示し，数多くの実験がなされたあとに，アンダーソンとストラウブによって曲がりによる損失水頭が求められるようになった．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_b = f_{b1} f_{b2} \cfrac{v^2}{2g} 
    \end{align}
$$

</span>

このとき，$h_b$は曲がりによる損失水頭[m]，$f_{b1}$は曲がり損失係数[単位なし]，$f_{b2}$は中心角が$90°$でないときの補正係数[単位なし]である．

それぞれの係数は下図によって求めることが出来る．

##### [2]屈折による損失水頭

屈折による損失水頭は屈折部で流れが剥離するために生じる．

ワイズバッハの実験式から次のように表される．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_{be} = f_{be} \cfrac{v^2}{2g} \\
        &f_{be} = 0.946sin^2 \cfrac{\theta}{2} + 2.05sin^4 \cfrac{\theta}{2}
    \end{align}
$$

</span>

このとき，$h_{be}$は屈折による損失水頭[m]，$f_{be}$は屈折損失係数[単位なし]である．

#### ④弁類などによる損失水頭

菅水路の途中に挿入された弁類は，スルース弁，バタフライ弁，コックなどがある．

弁類による損失係数は，バルブ自体の構造や開度によって異なってくる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_v = f_v \cfrac{v^2}{2g} 
    \end{align}
$$

</span>

このとき，$h_v$は弁類などによる損失水頭[m]，$f_v$は弁類などによる損失係数[単位なし]である．

#### ⑤流出による損失水頭

菅水路の末端において，水が下流側水槽や大気に放出させる場合，流れの持つ速度水頭がすべて失われる．

したがって，流出による損失水頭は次式で表され，流出損失係数は特別な場合を除き1.0と考えてよい．

<span style="font-size:150%;">
    
$$
    \begin{align}
        &h_0 = f_0 \cfrac{v^2}{2g} \\
        &f_0 = 1.0
    \end{align}
$$

</span>

このとき，$h_0$は流出による損失水頭[m]，$f_0$は流出損失係数[単位なし]である．

### 1.5.6. 単線菅水路

**単線菅水路**とは，管径や管種が異なっても，上流端から下流端まで１本の管で連結されている菅水路のことである．

では，単線菅水路での水位差と流量の式を求めていく．

まずは，水槽の入り口と出口の２点でベルヌーイの定理を適応させる．

また，水槽内の速度は$v_{01}$と$v_{02}$と置く．

<span style="font-size:150%;">
    
$$
    \begin{align}
        \alpha \cfrac{v_{01}^2}{2g} + z_1 + h_1 = \alpha \cfrac{v_{02}^2}{2g} + z_2 + h_2 + H
    \end{align}
$$

</span>

このとき，水槽内の速度$v_{01}$と$v_{02}$は管内の流速と比べるとものすごく遅いので$v_{01} = v_{02} = 0$となる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        z_1 + h_1 - (z_2 + h_2) = H
    \end{align}
$$

</span>

水位差$H$は管水路の入り口から出口までに生じる損失水頭の総和と等しくなる．

<span style="font-size:150%;">
    
$$
    \begin{align}
        H &= h_{l1} + h_{l2} + h_e + h_b + h_{be} + h_{se} + h_v + h_0 \\
          &= f_1 \cfrac{l_1}{D_1} \cfrac{v_1^2}{2g} + f_2 \cfrac{l_2}{D_2} \cfrac{v_2^2}{2g} + f_e \cfrac{v_1^2}{2g} + f_b \cfrac{v_1^2}{2g} + f_{be} \cfrac{v_1^2}{2g} + f_{se} \cfrac{v_1^2}{2g} + f_v \cfrac{v_2^2}{2g} + f_0 \cfrac{v_2^2}{2g} \\
          &= \left(f_1 \cfrac{l_1}{D_1} + f_e + f_b + f_{be} + f_{se}  \right) \cfrac{v_1^2}{2g} + \left(f_2 \cfrac{l_2}{D_2} + f_v + f_0 \right)\cfrac{v_2^2}{2g}
    \end{align}
$$

</span>
