## [Python]Anacondaで仮想環境を作る方法

###    1. minicondaをインストールする
miniconda path : https://docs.anaconda.com/free/miniconda/ 

###    2. Anaconda_Prompt
[仮想環境を作る](#仮想環境を作る) $ \quad$ [仮想環境の確認](#仮想環境の確認) $ \quad $[仮想環境の削除](#仮想環境の削除)　

[仮想環境をActiveにする](#仮想環境をActiveにする) $ \quad$ [仮想環境を非Activeにする](#仮想環境を非Activeにする) $ \quad $[パッケージの一覧を見る](#パッケージの一覧を見る)

[Jupyterを入れる](#Jupyterを入れる)

#### 仮想環境を作る

```
    conda create -n 仮想環境の名前 python パッケージ
```

#### 仮想環境の確認
```
    conda info -e
```
#### 仮想環境の削除

```
    conda remove -n 仮想環境の名前 --all
```
#### 仮想環境をActiveにする

```
    conda activate 仮想環境の名前
```
#### 仮想環境を非Activeにする

```
    conda deactivate
```
#### パッケージの一覧を見る

```
    conda list
```
#### Jupyterを入れる

```
    conda install jupyter notebook
```
###   3.Run (pynote)
```
    conda create -n pynote 
    conda activate pynote
    conda install jupyter notebook
    pip install pandas
    pip install matplotlib
    pip install sympy
    jupyter notebook

```

### lambda と def
lambdaは無名関数，defは宣言文

lambda x : x + 1

filter（関数，リストなどのオブジェクト）条件に合致するものを抽出する
filter(lambda num : num > 90, nums)

map（関数，リストなどのオブジェクト）リスtなどの複数要素を持つデータに対して，その要素に同じ処理を行う関数

map(lambda name : name + "さん", pre_list)
