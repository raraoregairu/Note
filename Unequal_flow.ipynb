# 現場のための水理学

## 流れの概念

## 不等流計算

### 不等流
>場所によって速さや深さが変わるような流れ

### 不定流
>時間的に流量が変化する

**一般的な河川の流れは時間的な変化を考えない<span style="color: red;">定流</span>とみなしても差し支えない**

### ある地点に水が流れる時の表現要素

 - 流体の置かれている位置の高さ
 - 水深
 - 流速　　　　

### 流れの表示
> 流れを構成する3つの要素は長さの次元，つまり水頭（エネルギー）をもって表現する．

### エネルギー線

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        E = z + h + V^2 / 2g \tag{1.1}
    \end{align}
$
    
</span>
<br>
<br>

#### 流れの表示



<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        &V ; 流速(m / s) \quad h; 水深(m) \quad z ; 河床高(m) \quad H ; 水位(m) \quad \\
        &i_{e} ; エネルギー勾配 \quad i_{w} ; 水面勾配 \quad i_{b} ; 河床勾配 \quad g ; 重力加速度(m / s^2)
    \end{align}
$
    
</span>
<br>
<br>

### 河床勾配 $i_{b}$
>河床の基準線（一般に水平）に対する傾き

### 水面勾配 $i_{w}$
>水面の基準線に対する傾き

### エネルギー勾配 $i_{e}$
>エネルギー線が基準線となす傾き

## 不等流計算

### 基礎式
不等流の場合，任意の二つの断面間で次式が成り立つ．

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        z_{1} + h_{1} + \cfrac{V_{1}^2}{2g} = z_{2} + h_{2} + \cfrac{V_{2}^2}{2g} + h_{f} \tag{2.1}
    \end{align}
$
    
</span>
<br>
<br>

#### 不等流の表示



<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        &V ; 流速(m / s) \quad h; 水深(m) \quad z ; 河床高(m) \quad H ; 水位(m) \quad h_{f} ; 摩擦損失水頭(m) \\
        &l ; 断面間距離 \quad i_{e} ; エネルギー勾配 \quad i_{w} ; 水面勾配 \quad i_{b} ; 河床勾配 \quad g ; 重力加速度(m / s^2)
    \end{align}
$
    
</span>

#### 摩擦損失水頭 $h_{f}$
>水が流下するとき，流水と川底がこすれたりして生じる摩擦などで失われるエネルギー

#### 一般化

断面間の距離を微小長さ$\varDelta x(m)$とする．

エネルギー勾配が$i_{e} = h_{f} / l$より$l$を$\varDelta x(m)$とおくと

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        h_{f} = i_{e} \cdot \varDelta x \tag{2.2}
    \end{align}
$
    
</span>
<br>
<br>

以上より式(2.1)を書き直すと

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        z_{1} + h_{1} + \cfrac{V_{1}^2}{2g} = z_{2} + h_{2} + \cfrac{V_{2}^2}{2g} + i_{e} \cdot \varDelta x　\tag{2.3}
    \end{align}
$
    
</span>
<br>
<br>

上式において，右辺から左辺を引いて，各項を$\varDelta x$で割ると

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{z_{2} - z_{1}}{\varDelta x} + \cfrac{h_{2} - h_{1}}{\varDelta x} + \cfrac {\cfrac{V_{2}^2}{2g} - \cfrac{V_{1}^2}{2g}}{\varDelta x} + i_{e} = 0　\tag{2.4}
    \end{align}
$
    
</span>
<br>
<br>

ここで，$\varDelta x(m)$をさらに微少量，つまり限りなく０に近い長さとして$dx$で表し，また，$z$や$h$，$V^2 / 2g$の変化量も$dz$や$dh$，$d{V^2 / 2g}$で表すと式(2.4)は

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dz}{dx} + \cfrac{dh}{dx} + \cfrac{d}{dx} \cfrac{V^{2}}{2g} + i_{e} = 0　\tag{2.5}
    \end{align}
$
    
</span>
<br>
<br>
また，水位$H$は，$H = h + z$であるので，

### 微分表示

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dH}{dx} + \cfrac{d}{dx} \left(\cfrac{V^{2}}{2g} \right) + i_{e} = 0　\tag{2.6}
    \end{align}
$
    
</span>
<br>
<br>

### 不等流の平均流速式（マニングの型の式）

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        V = \cfrac{1}{n} R^{2 / 3} i_{e}^{1 / 2}　\tag{2.7}
    \end{align}
$
    
</span>
<br>
<br>

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        V ; 平均流速(m / s) \quad n ; 粗度係数 \quad R ; 径深(m) \quad i_{e} ; エネルギー勾配
    \end{align}
$
    
</span>
<br>
<br>

#### 用語の定義



#### 粗度係数 $n$
>河床の粗さや形状などに左右される抵抗を表すもの

#### 径深 $R$
>流水断面積もしくは流積$A(m^2)$を潤辺（断面において水に潤っている辺の長さ）$S(m)$で割ったもの

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        R = A / S \tag{2.8}
    \end{align}
$
    
</span>
<br>
<br>

一般に大部分の河川のように水深に比して川幅が広い場合には，径深$R$は水深$h$に等しい．（広矩形断面）

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        R \simeq h \tag{2.9}
    \end{align}
$
    
</span>
<br>
<br>

式(2.7)より，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        i_{e} = \left( \cfrac{nV}{R^{2 / 3}} \right)^2 \tag{2.10}
    \end{align}
$
    
</span>
<br>
<br>
また，平均流速$V$は流量$Q(m^3 / s)$を流積$A(m^2)$で割ったものであるので，式(2.10)は，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        i_{e} = \left( \cfrac{nQ}{AR^{2 / 3}} \right)^2 = \cfrac{n^2 Q^2}{A^2 R^{4 /3}} \tag{2.11}
    \end{align}
$
    
</span>
<br>
<br>

これを式(2.6)に代入して，$V = Q / A$より

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dH}{dx} + \cfrac {1}{2g} \cfrac{d}{dx} \left(\cfrac{Q}{A} \right)^2 + \cfrac{n^2 Q^2}{A^2 R^{4 /3}} = 0 \tag{2.12}
    \end{align}
$
    
</span>
<br>
<br>

特に広矩形断面の場合，$R \simeq h$であり，流積$A$は，川幅$B$と水深$h$の積で表されることから，

### 広矩形断面を持つ水路の不等流計算

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dH}{dx} + \cfrac {1}{2g} \cfrac{d}{dx} \left(\cfrac{Q}{BH} \right)^2 + \cfrac{n^2 Q^2}{B^2 h^{10 /3}}  \tag{2.13}
    \end{align}
$
    
</span>
<br>
<br>

#### 流れに関する注意

広矩形断面の不等流について
$H$は水位であるので，$H = z + h$より式(2.13)の左辺の第一項は，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac {dH}{dx} = \cfrac{d(z + h)}{dx} = \cfrac{dz}{dx} + \cfrac{dh}{dx} \tag{2.14}
    \end{align}
$
    
</span>
<br>
<br>

また，流量$Q$は一定，川幅$B$も一定とすると，$ \cfrac {d}{dx} (h^{-2}) = \cfrac{- 2} {h^3} \cfrac{dh}{dx}$より，左辺第二項は，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac {1}{2g} \cfrac{d}{dx} \left(\cfrac{Q}{BH} \right)^2 = \cfrac {Q^2}{2gB^2} \cfrac{d}{dx}(h^{-2}) = - \cfrac{Q^2}{gB^2h^3} \cfrac{dh}{dx} \tag{2.15}
    \end{align}
$
    
</span>
<br>
<br>

式(2.14)，式(2.15)を式(2.13)に代入すると，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dz}{dx} + \cfrac{dh}{dx} - \cfrac{Q^2}{gB^2h^3} \cfrac{dh}{dx} + \cfrac{n^2 Q^2}{B^2 h ^{10 / 3}} = 0 \tag{2.16}
    \end{align}
$
    
</span>
<br>
<br>

$dh / dx$について整理して解くと

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dh}{dx} = \cfrac{- \cfrac{dz}{dx} - \cfrac{n^2 Q^2}{B^2 h^{10 / 3}}}{1 - \cfrac{Q^2}{gB^2h^3}} \tag{2.17}
    \end{align}
$
    
</span>
<br>
<br>

また，上流から下流に向かう方向を$x$の正の向きとすると，$-dz / dx = i_b$（河床勾配）となるから，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dh}{dx} = \cfrac{i_b - \cfrac{n^2 Q^2}{B^2 h^{10 / 3}}}{1 - \cfrac{Q^2}{gB^2h^3}} \tag{2.18}
    \end{align}
$
    
</span>
<br>
<br>

一般的な不等流の式を，水深の場所的変化（$dh / dx$）を表現する式となる．

式（2.18）について分子 = 0すなわち，$dh / dx = 0$の水深が場所的に変化しないような場合が考えられる．（等流水深）

### 等流水深 $h_{o}$

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        h_{o} = \left( \cfrac{n^2 Q^2}{B^2 i_{b}} \right)^{3 / 10} = \left( \cfrac{nQ}{B \sqrt{i_{b}}} \right)^{3 / 5} \tag{2.20}
    \end{align}
$
    
</span>

<br>
<br>

等流とはいたるところで水深の等しい特別な流れといえる．

#### 流れに関する注意

一方，式(2.18)について，分母 = 0，すなわち$dh / dx =  \pm \infty$となる場合が考えられる．（限界水深）すなわち，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        1 = \cfrac{Q^2}{g B^2 h^3} \tag{2.21}
    \end{align}
$
    
</span>
<br>
<br>

これを$h$について整理したものが限界水深$h_{c}$である．

### 限界水深 $h_{c}$

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        h_{o} = \sqrt[3]{\cfrac{Q^2}{gB^2}} \tag{2.22}
    \end{align}
$
    
</span>
<br>
<br>


$h > h_{c}$の流れを **常流**，$h = h_c$の流れを **限界流**，また，$h < h_c$の流れを **射流** と定義する．

#### 常流・射流の区分

<br>
<br>

|<span style="font-size:150%;"> </span>|<span style="font-size:150%;">常流・射流の区分</span>|
|---|-----------|
|<span style="font-size:200%;">常　流</span>|<span style="font-size:150%;">$h > h_c$, $F_r < 1$</span>|
|<span style="font-size:200%;">限界流</span>|<span style="font-size:150%;">$h = h_c$, $F_r = 1$</span>|
|<span style="font-size:200%;">常　流</span>|<span style="font-size:150%;">$h < h_c$, $F_r > 1$</span>|

<br>
<br>
<span style="font-size:120%;">
※ $h_c$ ; 限界水深 $F_r$ ; フルード数
</span>
<br>
<br>

ある水深（限界水深）を境に性質の異なる流れの形態が存在する．

<u>射流の場合，下流側で何らかの乱れが生じてもその影響を受けないが常流ではそのような影響が上流に及ぶ </u>

### 広矩形単断面の不等流計算

対象とする最小の計算区間を上流側断面１，下流側断面２の間に設定する．

#### 不等流の表示図



#### 差分表示について

式(2.13)を実際の計算に用いることができるように，諸量の差をもって表す．

<br>
<br>
<span style="font-size:120%;">
$
    \begin{align}
        \cfrac{H_2 - H_1}{\varDelta x} &+ \cfrac{Q^2}{2g \cdot \varDelta x} \left\{ \left( \cfrac{1}{B_2 h_2} \right)^2 - \left( \cfrac{1}{B_1 h_1} \right)^2 \right\} + \cfrac{n^2 Q^2}{2} \left\{ \left( \cfrac{1}{B_2^2 h_2^{10 / 3}} \right) + \left( \cfrac{1}{B_1^2 h_1^{10 / 3}} \right) \right\} = 0 \tag{2.25}
    \end{align}
$
    
</span>
<br>
<br>

ただし，流量$Q$は一定，第三項目は断面１での値と断面２での値の平均である．

ゆえに，両辺に<span style="font-size:80%;">$\varDelta x$</span>をかけ，左辺に下流側，右辺に上流側の諸量を示す項をもってくると，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \underbrace{ \left[ H_2 + \cfrac{Q^2}{2g B_2^2 h_2^2} + \cfrac{n^2 Q^2 \varDelta x}{2 B_2^2 h_2^{10 / 3}} \right]}_{下流側} = \underbrace{ \left[ H_1 + \cfrac{Q^2}{2g B_1^2 h_1^2} + \cfrac{n^2 Q^2 \varDelta x}{2 B_1^2 h_1^{10 / 3}} \right]}_{上流側}  \tag{2.26}
    \end{align}
$
    
</span>
<br>
<br>

また，水位$H$を河床高$z$と水深$h$の和で表すと，

### 差分表示

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \underbrace{ \left[ z_2 + h_2 + \cfrac{Q^2}{2g B_2^2 h_2^2} + \cfrac{n^2 Q^2 \varDelta x}{2 B_2^2 h_2^{10 / 3}} \right]}_{下流側} = \\ \underbrace{ \left[ z_1 + h_1 + \cfrac{Q^2}{2g B_1^2 h_1^2} + \cfrac{n^2 Q^2 \varDelta x}{2 B_1^2 h_1^{10 / 3}} \right]}_{上流側}  \tag{2.27}
    \end{align}
$
    
</span>
<br>
<br>

式(2.27)のなかで，流量$Q$，河床高$z$，川幅$B$，断面間の距離$\varDelta x$は既知量であり，重力加速度$g$や粗度係数$n$は定数．

よって，水深$h$が未知量として求めるべきものとなる．

<br>

常流の場合，流れの変化は下流から上流に及ぶため，下流側の$h_2$が既知量として与えられ，上流側の$h_1$を求める．（下流端の水深が境界条件として与えられる．）

一方，射流では，流れは上流から下流に変化するため，上流側の$h_1$が既知量として与えられ，下流側の$h_2$を求める．（上流端の水深が境界条件として与えられる．）


### 一般断面の不等流計算

潤辺の近くに多くの凹凸がある場合，基礎式は

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{dH}{dx} + \cfrac{d}{dx} \left( \alpha \; \cfrac{V^2}{2g} \right) + i_e = 0 \tag{2.28}
    \end{align}
$
    
</span>
<br>
<br>

$\alpha$ はエネルギー補正係数であり，定義づけが必要である．また，平均流速$V$やエネルギー勾配$i_e$について凹凸のある全断面の平均値なので，単断面のように簡単に導くことができない．

**そこで，下記のように断面を小さな矩形のブロックに分けて考える．**


#### 一般断面形



<br>
<br>
<span style="font-size:150%;">
$h_i$ ; 第$i$ブロックの水深 $ \quad \varDelta y_i$ ; 第$i$ブロックの断面幅 $\quad n_i$ ; 第$i$ブロックの粗度係数 $\quad z_i$ ; 第$i$ブロックの河床高  
</span>
<br>
<br>

全断面を幅$\varDelta y$で$N$個の細かいブロックに分割し，第$i$ブロックの諸量について考える．断面積$\varDelta a_i$は

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \varDelta a_i = h_i \cdot \varDelta y_i \tag{2.29}
    \end{align}
$
    
</span>
<br>
<br>

また，分担流量は

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \varDelta q_i = u_i \varDelta a_i = u_i h_i \varDelta y_i \tag{2.30}
    \end{align}
$
    
</span>
<br>
<br>

ここで，$u_i$は$i$ブロックにおける平均流速であり，マニングの公式（式（2.7））より

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \varDelta u_i = \cfrac{1}{n_i} r_i^{2 / 3} i_e^{1 / 2} \tag{2.31}
    \end{align}
$
    
</span>
<br>
<br>

$i_e$はエネルギー勾配である．
$r_i$は$i$ブロックの径深であり，断面積を潤辺で除して求められるが，ブロック分けした細かい矩形断面の潤辺は，断面幅$\varDelta y_i$に近似できることから

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \varDelta r_i \fallingdotseq \varDelta a_i / \varDelta y_i = h_i \tag{2.32}
    \end{align}
$
    
</span>
<br>
<br>

これを式（2.31）に代入すると，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        u_i = \cfrac{1}{n_i} h_i^{2 / 3} i_e^{1 / 2} \tag{2.33}
    \end{align}
$
    
</span>
<br>
<br>

ゆえに式（2.30）は

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \varDelta q_i = \left( \cfrac{1}{n_i} h_i^{2 / 3} i_e^{1 / 2} \right) \left( h_i \varDelta y_i \right) = \left( \cfrac{1}{n_i} h_i^{5 / 3} \varDelta y_i \right) i_e^{1 /2} \tag{2.34}
    \end{align}
$
    
</span>
<br>
<br>

上式の総和が断面における全流量となることから，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        Q = \displaystyle \sum_{i = 1}^N \varDelta q_i = \displaystyle \sum_{i = 1}^N \left( \cfrac{1}{n_i} h_i^{5 / 3} \varDelta y_i \right) i_e^{1 / 2} \tag{2.35}
    \end{align}
$
    
</span>
<br>
<br>

ここで，エネルギー勾配$i_e$は断面のいたるところで近似的に一定とみなせることから，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        i_e = \cfrac{Q^2}{\left( \displaystyle \sum_{i = 1}^N \cfrac{1}{n_i} h_i^{5 / 3} \varDelta y_i \right)^2 } \tag{2.36}
    \end{align}
$
    
</span>
<br>
<br>

また，エネルギー補正係数$\alpha$は，$V = Q / A$より

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \alpha = \cfrac{\displaystyle \sum_{i = 1}^N u_i^3 \varDelta a_i}{V^3 A} = \cfrac{\displaystyle \sum_{i = 1}^N u_i^3 \varDelta a_i}{ \left( Q^3 /A^2 \right)} \tag{2.37}
    \end{align}
$
    
</span>
<br>
<br>

    \end{align}
$
    
</span>

## 掃流砂と河床変動

### 掃流力（河床せん断力） $\tau_0$

河床上をある水深で水が流れると，河床の潤辺には単位面積当たりの$\tau_0$のせん断力が水塊に対して流れと逆方向に働き，

一方，流れは潤辺に対して同方向に$\tau_0$の力を及ぼす．

したがって，河床が非粘着性の砂礫で構成されている場合には，この力は砂礫粒子を下流へ押し流そうとする．

#### 掃流力の定義



河床勾配$i_b$の河川に等流状態で水が流れるとき，流下方向に長さ$l$の区間を考える．

$l$区間の水塊の容積$V_t$は，流積を$A$とすると，

<span style="font-size:150%;">
$
    \begin{align}
        V_t = Al
    \end{align}
$
    
</span>
<br>
<br>

$l$区間の水塊の重量$W$は，水の密度を$\rho$とすると,

<span style="font-size:150%;">
$
    \begin{align}
        W = \rho g V_t = \rho g A l
    \end{align}
$
    
</span>
<br>
<br>

したがって，$W$の斜面方向の成分は

<span style="font-size:150%;">
    
$
    \begin{align}
        W_x = W sin i_b \fallingdotseq W i_b = \rho g A l i_b
    \end{align}
$
    
</span>
<br>
<br>

一方，$l$区間の潤辺$S$に加わる全せん断力$T$は

<span style="font-size:150%;">
$
    \begin{align}
        T = \tau_0 S l 
    \end{align}
$
    
</span>
<br>
<br>

したがって，等流状態では流速は一定であるから，$W_x = T$，すなわち，

<span style="font-size:150%;">
$
    \begin{align}
        \rho g A l i_b = \tau_0 S l
    \end{align}
$
    
</span>
<br>
<br>

となり，掃流力$\tau_0$は径深を$R (= A / S)$として，

### 掃流力

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \tau_0 = \rho g \cfrac{A}{S} i_b = \rho g R i_b \tag{3.1}
    \end{align}
$
    
</span>
<br>
<br>

式(3.1)は不等流においても，河床勾配$i_b$の代わりにエネルギー勾配$i_e$を用いると近似的に表すことができる．

<span style="font-size:150%;">
$
    \begin{align}
        \tau_0 = \rho g R i_e \tag{3.2}
    \end{align}
$
    
</span>
<br>
<br>

ただし，$i_e = \cfrac{d}{dx} \left( \cfrac{\alpha V^2}{2g} + H \right),H : $基準水平面からの水位 $V : $平均流速，$ \alpha : $エネルギー補正係数，$x : $流下方向の座標軸

### 摩擦速度
> 掃流力$\tau_0$を流速の次元で定義したもの

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        u_* = \sqrt { \tau_0 / \rho } = \sqrt{g R i_e} \tag{3.3}
    \end{align}
$
    
</span>
<br>
<br>

したがって，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \tau_0 = \rho u_*^2 \tag{3.4}
    \end{align}
$
    
</span>
<br>
<br>

### 無次元掃流力（無次元せん断力） $ \tau_*$
> 摩擦速度$u_*$を無次元表示したもの

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \tau_* = \cfrac{u_*^2}{sgd} = \cfrac{R i_e}{sd} \tag{3.5}
    \end{align}
$
    
</span>
<br>
<br>

ただし，$s : $砂粒の水中比重$s = \cfrac{\rho_s - \rho}{\rho}$，$\rho_s : $砂粒の密度，$d : $砂の粒径，$i_e : $エネルギー勾配

### 限界掃流力 $\tau_c$ あるいは，限界摩擦速度 $u_{*c}$
>ある粒子が河床上にあるとき，河床における掃流力$\tau_0$あるいは，摩擦速度$u_*$がある限界値を超えると粒子は移動を開始する．その限界値．

粒子が移動するかは次の不等式で表されえる

$\qquad $静止 ･･････ $u_* < u_{*c}$ あるいは $\tau_0 < \tau_c$

$\qquad $移動 ･･････ $u_* > u_{*c}$ あるいは $\tau_0 > \tau_c$

### 無次元限界掃流力
>限界摩擦速度$u_{*c}$を摩擦速度と同様に無次元表示にしたもの

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \tau_{*c} = \cfrac{u_{*c}^2}{sgd} \tag{3.6}
    \end{align}
$
    
</span>
<br>
<br>

### 岩垣の式

限界摩擦速度$u_{*c}$を算出する方法について一様粒径砂礫の限界掃流力の算定式のひとつ．

砂粒の水中比重$s \fallingdotseq 1.65$，動粘性係数 $\nu \fallingdotseq 0.01 cm^2 /s(20.3℃)$，重力加速度$g = 980 cm / s^2$で表すと次式となる．（cm-sec 単位）

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \qquad &d \geqq 0.303 \hspace{0.3pc}cm ; u_{*c}^2 = 80.9d \\
        0.118 \leqq &d \leqq 0.303 \hspace{0.3pc}cm ;\hspace{1.3pc} = 134.6d^{31 / 22} \\
        0.0565 \leqq &d \leqq 0.118 \hspace{0.3pc}cm ; \hspace{1.3pc} = 55.0d  \tag{3.7}\\
        0.0065 \leqq &d \leqq 0.0565 cm ; \hspace{1.3pc} = 8.41d^{11 / 32} \\
        \qquad &d \leqq 0.0065 cm ; \hspace{1.3pc} = 226d \\
    \end{align}
$
    
</span>
<br>

## 掃流砂量式

水の流れによって河床を構成する砂礫が移動する場合に移動形式は大別して掃流と浮遊の二つがある．

### 掃流砂
> 砂礫が流水の流れ方向の抵抗力によって河床付近を移動する流砂．実際の河川において，河床変動に対して支配的な影響を持つ．

### 浮遊砂
> 流水の乱れによる拡散作用によって上方に巻き上げられ，流路断面内を流水とともに移動する流砂．粒径が細かい河口部やダムにおける堆砂問題の場合に影響を持つ．

<br>

粒子の沈降速度を$w_f$，流水の流れによる上方への粒子の移動速度を$u_s$とすると，粒子が浮遊するか沈降するかについて次の不等式が成立する．

    浮遊 ･･････ $u_s > w_f$
    沈降 ･･････ $u_s < w_f$

一般に，$u_s$は$u_*$に比例する．

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        u_s = 0.6u_*～ 0.93u_* \tag{4.1}
    \end{align}
$
    
</span>
<br>
<br>

浮遊限界は$u_s / w_f = 1$であるから，(4.1)式より

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        \cfrac{u_s}{w_f} = \cfrac{0.6u_*}{w_f} ～ \cfrac{0.93u_*}{w_f} = 1 \tag{4.2}
    \end{align}
$
    
</span>
<br>
<br>

すなわち，

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        1.08 < \cfrac{u_*}{w_f} < 1.67 \tag{4.2}
    \end{align}
$
    
</span>
<br>
<br>

したがって，掃流と浮遊の卓越領域は$u_*$と$w_f$を用いると次式で表すことができる．

### 掃流卓越領域

<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        \cfrac{u_*}{w_f} < 1.08 \tag{4.3}
    \end{align}
$
    
</span>
<br>

### 掃流・浮遊の混在領域

<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        1.08 < \cfrac{u_*}{w_f} < 1.67 \tag{4.4}
    \end{align}
$
    
</span>
<br>

### 浮遊卓越領域

<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        1.67 < \cfrac{u_*}{w_f}  \tag{4.5}
    \end{align}
$
    
</span>
<br>

ここで，沈降速度$w_f$の式として，Rubyの実験式がよく用いられる．

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        \cfrac{w_f}{\sqrt{sgd}} = \sqrt{\cfrac{2}{3} + \cfrac{36 \upsilon^2}{sgd^3}} - \sqrt{\cfrac{36 \upsilon^2}{sgd^3}} \tag{4.6}
    \end{align}
$
    
</span>
<br>
<br>

ただし，$\upsilon :$水の動粘性係（$\fallingdotseq 0.01 cm^2 / s$），$s :$砂粒の水中比重（砂の比重1.65），$d :$砂粒の粒径（cm），$g :$重力加速度（$980cm / s^2$）

流砂量式のほとんどが一様砂礫を対象とした取り扱いにって誘導されているために，混合砂礫河床からなる実際河川への適用には注意が必要である．

また，実際河川では，河床波の発生，変化に伴って流れの抵抗も変化するが，これを考慮した計算法はまだ確立されていない．

断面内の掃流砂量を$Q_s（m^3 / s）$，単位幅あたりの掃流砂量を$q_B（m^2 / s）$と書き表すと，一般に$q_B$は無次元掃流砂$\tau_*$あるいは摩擦速度$u_*$の関数，すなわち，$q_B = f(\tau_*)$あるいは$q_B = f(u_*)$として与えられる．

### 佐藤・吉川・芦田の式（土研公式）

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        q_B = \cfrac{u_*^3}{sg} F \left(\cfrac{u_*^2}{u_{*c}^2}\right)f(n) \tag{4.7}
    \end{align}
$
    
</span>
<br>
<br>

ここで，Manningの粗度係数を$n$として，$n \geqq 0.025$のとき$f(n) = 0.623$，$n \leqq 0.025$のとき，$f(n) = 0.623(40n)^{-3.5}$である．

または，$F(u_*^2 / u_{*c}^2)$の近似式として次式がある．

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        F\left(\cfrac{u_*^2}{u_{*c}^2}\right) = \cfrac{1}{1 + 8(u_*^2 / u_{*c}^2)^4} \tag{4.8}
    \end{align}
$
    
</span>
<br>
<br>

### Meyer-Peter・Mullerの式

<br>
<br>
<span style="font-size:150%;">
    
$
    \begin{align}
        q_B = 8(\tau_*' - \tau_{*c})^{1.5} \sqrt{sgd^3} \tag{4.9}
    \end{align}
$
    
</span>
<br>
<br>

<br>
<br>
<span style="font-size:150%;">

    \end{align}
$$
    
</span>
<br>
<br>

ただし，$u_*' = (n_b / n)^{3/4} u_* :$有効摩擦速度，$n_b = 0.0192d_{90}^{1 / 6}(d_{90} : cm単位)$：砂粒抵抗を表わすStlickler型の粗度係数，$n :$流れ全体のManningの粗度係数

> 河床に河床波がある場合に，それは流れの抵抗に対して形状抵抗として寄与するものと考え，形状抵抗として分離した残りの砂面上の表面抵抗分を有効掃流力と定義する．

## 河床変動の１次元解析法

通常単に河床変動といえば，河道の数kmから数10kmにわたる大規模な河床変動をさしており，その計算には河床を横断方向に平均化し，縦断方向のみの変化について論ずる１次元解析法が適用される．

例えば，ダム築造や河川改修によって河川の上・下流の境界条件が変化する場合に，河床変動を長期間，かつ広範囲にわたって予測するのにこの方法は有効である．

一様砂礫あるいは平均粒径を用いた掃流砂による河床変動の１次元解析法の基礎式は以下の式である．

流れの運動方程式と連続式；

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{\partial z}{\partial x} + \cfrac{\partial h}{\partial x} + \cfrac{\partial}{\partial x}\left(\cfrac{\alpha V^2}{2g}\right) + i_e = 0 \tag{5.1}
    \end{align}
$
    
</span>
<br>
<br>

<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{\partial}{\partial x}(BhV) = 0 \tag{5.2}
    \end{align}
$
    
</span>
<br>
<br>

流砂の連続式と流砂量式；

<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{\partial z}{\partial t} + \cfrac{1}{(1-\lambda)B} \cdot \cfrac{\partial}{\partial x}(q_BB)= 0 \tag{5.3}
    \end{align}
$
    
</span>

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        q_B = f(u_*) \tag{5.4}
    \end{align}
$
    
</span>
<br>
<br>

ただし，$B :$川幅，$h:$平均水深，$V:$平均流速，$x:$流下方向の座標軸，$z:$平均河床高，$t:$時間，$u_*:$平均摩擦速度，$i_e:$エネルギー勾配，$q_s:$掃流砂量，$\lambda:$河床砂の空隙率，$\alpha:$エネルギー補正係数

### 流砂の連続式

河川流路の上・下流に任意の２断面を考えると，その上流測断面①の流砂量が下流測断面②より大きければ，この区間には堆積が起き，逆に小さければ，洗堀が起こる．

このように，河川流路における河床高の変化は，各断面流砂量の不均衡によって生ずるものであり，掃流砂による河床高$z$の時間変化は，(5.3)式の連続式により求めることができる．

断面①，②の掃流砂量をおのおの$Q_{B1}, Q_{B2}$（容積表示，$m^2/s$）とすると，$\varDelta t$秒間に①断面を通過する流砂量は$Q_{B1} \cdot \varDelta t(m^3)$であり，同様に②断面では$Q_{B2} \cdot \varDelta t(m^3)$である．

したがって，この区間で河床に堆積と洗堀のいずれが起こるかは，河床変動高を$\varDelta z(m)$として以下のように表現できる．ただし，$z$軸は上向きを正とする．

<span style="font-size:120%;">
$
    \begin{align}
        Q_{B1} \cdot \varDelta t > Q_{B2} \cdot \varDelta tのとき，堆積する．(\varDelta z > 0)
    \end{align}
$
    
</span>
<br>

<span style="font-size:120%;">
$
    \begin{align}
        Q_{B1} \cdot \varDelta t = Q_{B2} \cdot \varDelta tのとき，堆積も洗堀もしない．(\varDelta z = 0)
    \end{align}
$
    
</span>
<br>

<span style="font-size:120%;">
$
    \begin{align}
        Q_{B1} \cdot \varDelta t < Q_{B2} \cdot \varDelta tのとき，洗堀する．(\varDelta z < 0)
    \end{align}
$
    
</span>
<br>
<br>

言い換えると，

<span style="font-size:120%;">
$
    \begin{align}
        Q_{B1} - Q_{B2} > 0 のとき，\varDelta z > 0
    \end{align}
$
    
</span>

<br>

<span style="font-size:120%;">
$
    \begin{align}
        Q_{B1} - Q_{B2} = 0 のとき，\varDelta z = 0
    \end{align}
$
    
</span>
<br>

<span style="font-size:120%;">
$
    \begin{align}
        Q_{B1} - Q_{B2} < 0 のとき，\varDelta z < 0
    \end{align}
$
    
</span>
<br>

河床に堆積が起こる場合，区間に堆積する量は$(Q_{B1} - Q_{B2}) \cdot \varDelta t$であるが，これを河床高に換算するときに空隙を考慮する必要がある．

空隙率を$\lambda$で表わすことにすると，堆積後の容積を$V_x$として次式のように表わされる．

<br>

<span style="font-size:150%;">
$$
    \begin{align}
       \cfrac{1 - \lambda}{1} = \cfrac{(Q_{B1} - Q_{B2}) \cdot \varDelta t}{V_x}
    \end{align}
$$
    
</span>
<br>

すなわち，

<br>

<span style="font-size:150%;">
$$
    \begin{align}
       V_x = \cfrac{Q_{B1} - Q_{B2}}{1 - \lambda}\cdot \varDelta t \tag{5.5}
    \end{align}
$$
    
</span>
<br>

一方，$V_x$を河床変動高$\varDelta z$で表わすと，区間距離を$\varDelta x$として，

<br>

<span style="font-size:150%;">
$$
    \begin{align}
       V_x = B \cdot \varDelta x \cdot \varDelta z \tag{5.6}
    \end{align}
$$
    
</span>
<br>

であるから，(5.5)式と(5.6)式より

<br>

<span style="font-size:150%;">
$$
    \begin{align}
       V_x = B \cdot \varDelta x \cdot \varDelta z \tag{5.6}
    \end{align}
$$
    
</span>

すなわち，
<br>

<span style="font-size:150%;">
$$
    \begin{align}
       \cfrac{\varDelta z}{\varDelta t} + \cfrac{1}{(1 - \lambda)B}\cdot \cfrac{Q_{B2} - Q_{B1}}{\varDelta x} = 0 \tag{5.7}
    \end{align}
$$
    
</span>
<br>

(5.7)式は距離に関して後進差分で表現しており，これを微分方程式で表わすと，

<br>

<span style="font-size:150%;">
$$
    \begin{align}
       \cfrac{\partial z}{\partial t} + \cfrac{1}{(1 - \lambda)B}\cdot \cfrac{\partial Q_B}{\partial x} = 0 \tag{5.8}
    \end{align}
$$
    
</span>
<br>

または，

<br>

<span style="font-size:150%;">
$$
    \begin{align}
       \cfrac{\partial z}{\partial t} + \cfrac{1}{(1 - \lambda)B}\cdot \cfrac{\partial (q_B \cdot B)}{\partial x} = 0 \tag{5.8}
    \end{align}
$$
    
</span>
<br>

なお，$\lambda$は砂の場合，0.4前後の値である．

## 河床変動の数値計算法

河床変動の数値計算法としては，最近は基礎式を直接差分化して逐次計算を行う，いわゆる差分法が一般であり，一様砂礫河床で掃流砂を対象とした場合の手順は以下のとおりである．

 1. 現河床形について，与えられた流量条件のもとに不等流計算を行い，掃流力の縦断分布を求める．
 2. 掃流砂量式を用いて流砂量の縦断分布を求める．
 3. 流砂の連続式により河床変動量を求める．

以下のプロセスを繰り返すことで，河床高の時間変化を予測することができる．

差分式による解法では，収束性，安定性，精度などを考慮されなければならない．特に安定性が満足されないと，解は発散して計算不能となる．

差分法により安定な数値解を得るために，差分スキームの与え方，差分幅$\varDelta t, \varDelta x$のとり方および境界条件の設定に注意を要する．

有限な伝播速度を有する現象では，実際の攪乱の伝播方向，すなわち特性曲線の方向と差分スキームによって規定される計算上の変動の伝播方向を一致させる必要がある．

河床変動について考えると，河床の微小攪乱の伝播方向は常流$(F_r < 1)$の場合には流下方向に対して正，射流$(F_r > 1)$の場合には負となる．ここで，フルード数$F_r = V / \sqrt{gh}$である．

したがって，<u>差分スキームのとり方は，常流の場合は時間に関して前進差分，距離に関して後進差分であり，射流の場合は，時間，距離に関していずれも前進差分としなければならない．</u>

# 不等流計算プログラム

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
df_m = pd.read_excel(path, sheet_name = ["set", "setd", "h", "z", "B", "q", "n"], indez_col = 0)
```

```python
def futoryu(df):
    df1 = df.copy()
    s = df1["set"]["val"]
    sd, danmen, g, flow, border, method , h, z, B, n, q = df1["setd"], s[0], s[1], s[2], s[3], s[4], df1["h"], df1["z"], df1["B"], df1["n"], df1["q"]
    H, Q, s, k, sumie = sym.Symbol("H"), sd["Q"], sd["s"], sd["k"], 0 
    sd.loc[1, "HH"] = h[1][1] + z[1][1]
    
    for p in range(s[1]):
        
        p1 = p + 1
        q.loc[1, p1] = B[p1][1] * (h[p1][1] ** (5 / 3)) * (ie ** (1 / 2)) / n[p1][1]
        ieq = h[p1][1] ** (5 / 3) * B[p1][1] / n[p1][1]
        sumie = sumie + ieq
        
    ie = (Q[1] ** 2) / (sumie ** 2)
        
    for j in range(danmen - 1):
        
        j1, j2 = j + 1, j + 2
        dx ,Q1, Q2 = k[j2] - k[j1], Q[j1], Q[j2]
        H1 = h[1][j1] + z[1][j1]
        H2 = H1 + z[1][j2] - z[1][j1]

        hfai, sumfai1, sumpusai1, sumfai2, sumpusai2, o= 0, 0, 0, 0, 0, 1

        for r in range(s[j1]):
            
            r1 = r + 1
            fai1, fai2 = (h[r1][j1] ** 3) * B[r1][j1] / (n[r1][j1] ** 3), ((H - z[r1][j2]) ** 3) * B[r1][j2] / (n[r1][j2] ** 3)
            pusai1, pusai2 = h[r1][j1] ** (5 / 3) * B[r1][j1] / n[r1][j1], ((H - z[r1][j2]) ** (5 / 3)) * B[r1][j2] / n[r1][j2]
            sumfai1, sumfai2 = sumfai1 + fai1, sumfai2 + fai2
            sumpusai1, sumpusai2 = sumpusai1 + pusai1, sumpusai2 + pusai2

        aa, bb1 = (Q2 ** 2) / (2 * g), dx * (Q2 ** 2) / 2
        cc1, cc2 = H1 + (Q1 ** 2) * sumfai1 / (2 * g * (sumpusai ** 3)), (Q1 ** 2) * dx / (2 * (sumpusai ** 2))
        
        if flow == "j":
            
            bb = (- 1) * bb1
            cc = cc1 + cc2

        else:
            
            bb = bb1
            cc = cc1 - cc2

        expr = H + aa * sumfai2 / (sumpusai2 ** 3) + bb / (sumpusai2 ** 2) - cc
        
        while True:
            
            fh1 = expr.subs(H, H2)

            if abs(fh1) < border: 
                
                sd.loc[j2, "HH"], sd.loc[j2, "nn"]= H2, int(o)
                sumie = 0
                
                for t in range(s[j2]):
                    
                    t1 = t + 1
                    h.loc[j2, t1] = H2 - z[t1][j2]
                    q.loc[j2, t1] = B[t1][j2] * (h[t1][j2] ** (5 / 3)) * (ie ** (1 / 2)) / n[t1][[j2]]
                    ieq = h[t1][j2] ** (5 / 3) * B[t1][j2] / n[t1][j2]
                    sumie = sumie + ieq   
                    
                ie = (Q1 ** 2) / (sumie ** 2)
                
                break

            else:
                
                if method == "b1":
                    
                    fbh1 = sym.diff(expr).subs(H, H2)
                    dh = (-1) * fh1 / fbh1
                    H2 = H2 + dh
                    
                elif method == "b2":
                    
                    fbh1 = sym.diff(expr).subs(H, H2)
                    fbh2 = sym.diff(sym.diff(expr)).subs(H, H2)
                    dh1 = np.power(fbh1 ** 2 - 2 * fh1 * fbh2, 1 / 2)
                    
                    if flow == "j":
                        
                        dh = ((-1) * fbh1 + dh1) / fbh2
                    else:
                        
                        dh = ((-1) * fbh1 - dh1) / fbh2
                        
                    H2 = H2 + dh

                else:
                    
                    H2 = H2 - 0.7 * fh1

                o = o + 1
                
                if o > 20:
                    
                    print("error")
                    break
                    
    return df1
```

```python
df1 = futoryu(df_m)

fig, ax = plt.subplots()
ax.plot(df1["setd"]["k"], df1["setd"]["HH"], label = "suisin")
ax.plot(df1["setd"]["k"], df1["z"][1], label = "kasyodaka")
ax.legend()
plt.show 
```
