その中でも、Markdownセルは、文章や見出し、リスト、リンクなどを記述するために使う．

Markdownセルを使うことで、ノートブック内で簡単に整ったテキストを表示できる．

<h2 style="border-bottom: 3px solid #3498db; display: inline-block; padding-bottom: 2px;">
📌 Markdownセルで使える記法
</h2>

Markdownセルで使える記法は以下のとおりである．

 - Markdown記法

 - HTML記法

 - LaTeX記法（数式など）
 
 これらを組み合わせて、テキスト、数式、HTML要素などをノートに追加できる．

<h3 style="border-bottom: 2px solid #3498db; display: inline-block; padding-bottom: 3px;">
📌 Markdown記法
</h3>

Markdownは，文書をシンプルに記述するために使われる軽量マークアップ言語である。

<br>

テキストを簡単に構造化できるように設計されており，コードを書く際に複雑なタグを使うことなく<br>基本的なフォーマットを適用することができる．

<details>
    <summary style="font-size: 80%; font-weight: bold;">▶ マークアップ言語とは</summary>
    <div style="border-left: 4px solid #3498db; padding: 8px; background-color: #f8f9fa;">
        Webサイトの文章・文字のサイズ・画像などにタグ付けをし，コンピュータに伝達するための言語．<br>
        「タグ」という特定の文字列で囲むことで，元の文に構造・装飾などを反映させる言語．
    </div>
</details>

<details>
    <summary style="font-size: 80%; font-weight: bold;">▶ タグとは</summary>
    <div style="border-left: 4px solid #3498db; padding: 8px; background-color: #f8f9fa;">
        ウェブページの構造を定義するために使用される要素．<br>各タグは&lt;タグ名&gt;内容&lt;/タグ名&gt;のような形で記述され，文章の見た目や機能を指定する．
    </div>
</details>

<h3 style="border-bottom: 1px solid #3498db; display: inline-block; padding-bottom: 2px;">
📝 特徴
</h3>

1. **手軽に** 文章構造を明示できる  
2. **簡単で、直感的に** 記述できる  
3. **特別なツール不要** で読み書き可能  
4. **対応アプリを使えば** 快適に文書作成できる  

<h3 style="border-bottom: 3px solid #3498db; display: inline-block; padding-bottom: 3px;">
📌 HTML記法
</h3>

HTMLは，ウェブページを作成するために使われる標準的なマークアップ言語である．

基本的には，タグを使って要素を囲む形で記述する．

HTMLでは，タグ内に「属性」を指定することができる．

属性はタグに追加情報を与え，要素の表示方法や動作を変更するために使われる．

<h3 style="border-bottom: 1px solid #3498db; display: inline-block; padding-bottom: 2px;">
📝 特徴
</h3>

1. **詳細に**文章構造を指定できる 
2. **柔軟で，多彩なデザインやレイアウト**を実現できる
3. **Webブラウザ**があればそのまま表示できる 
4. **CSSやJavaScriptと組み合わせれば**高度な表現が可能

```html
<font color="red" size="1">テキスト</font>  → <font color="red" size="1">フォント変更</font>
<span style="font-size: 200%; color:#00ff00;">テキスト</span> → <span style="font-size: 200%; color:#00ff00;">フォント変更</span>
```

🔹 `<font>` タグは非推奨のため、現在は `<span>` に style を適用する方法が一般的である．

<h4 style="border-bottom: 1px solid #3498db;display: inline-block; padding-bottom: 2px;">🛠 フォントを太字にする</h4>

<h5 style="display: inline-block; padding-bottom: 2px;">✏️ Markdownでの記述方法</h5>

```markdown
**テキスト** → **太字**
__テキスト__ → __太字__
```

🔹 `**` と `__` は同じ意味だが、`__` は文中の一部を強調する際に使われることが多い.

<details><summary><strong>💡 使用例</strong></summary>

```markdown
**これは太字です**
__これも太字です__
```

💻 実行結果

**これは太字です**  
__これも太字です__

</details>

<h5 style="display: inline-block; padding-bottom: 2px;">🌐 HTMLでの記述方法</h5>

```html
<strong>テキスト</strong> → <strong>太字</strong>
<b>テキスト</b> → <b>太字</b>
```

🔹 `<strong>` は **意味的な強調** を持つが、`<b>` は単なる視覚的な装飾である．

<details>
    <summary style="font-size: 90%; font-weight: bold;">▶ ホバー時とは</summary>
    <div style="border-left: 4px solid #3498db; padding: 8px; background-color: #f8f9fa;">
        ユーザーがマウスカーソルを特定の要素（リンク,ボタンなどに乗せたときに発生する動作．
    </div>
</details>

<details>
    <summary style="font-size: 90%; font-weight: bold;">▶ ツールチップとは</summary>
    <div style="border-left: 4px solid #3498db; padding: 8px; background-color: #f8f9fa;">
        ユーザーがマウスカーソルを特定の要素に合わせたときに表示される小さな情報ボックス．
    </div>
</details>

<details><summary><strong>💡 使用例</strong></summary>

```html
<strong>これは太字です</strong>
<b>これも太字です</b>
```

💻 実行結果

<strong lang="ja">これは太字です</strong>  
<b>これも太字です</b>

[テキスト1](#テキスト1)
```

<span style="font-size: 80%;">▶ 実行</span>

[テキスト1](#テキスト1)

</details>

---

参考

https://qiita.com/hennin/items/7ee58dd7d7c013a23be7

### 外部リンク

---

#### ◎Markdown : 

    []（角括弧）の中に表示したいテキスト，そのあと(URL)
    
<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```

```

<span style="font-size: 80%;">▶ 実行</span>

[Google検索エンジン](https://www.google.com/)

</details>

---

### 定義参照リンク

---

#### ◎Markdown : 

    同じリンクの参照を何度も利用する場合は、リンク先への参照を定義することができる．
    [表示するテキスト][定義するテキスト]，別の場所で[定義するテキスト]: 定義するリンク 

<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```
[テキスト][テキスト1]
[こっちでも][テキスト1]
    
[テキスト1]: リンク
```

<span style="font-size: 80%;">▶ 実行</span>

[参照先へ][テキスト1]
    
[テキスト1]: https://qiita.com/tbpgr/items/989c6badefff69377da7

</details>

---

### 画像を表示

---

#### ◎Markdown : 

    []（角括弧）の中に表示したいテキスト，そのあと(画像URL)
    
<details><summary><span style="font-size: 80%;">▶ 例</span></summary>

```

テーブル
「|」は列の区切りに使われる

空白を入れずに「|」入れることでその左のセルを結合することができる

（横にセルを連結することはできるが、縦に連結することはできない。）

複数の「-」はヘッダー部とボディ部の区切りに使われる。また、その両端を「:」で囲むことでその列を中央寄せに、右端に「:」を付けることで右寄せになる。

最後に「[]」で囲んだ文字列はテーブルのタイトルになる。
